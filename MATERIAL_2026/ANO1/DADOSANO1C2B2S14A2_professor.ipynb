## Base Teórica

### Exemplo de Leitura de Arquivo

Em Ciência de Dados, frequentemente precisamos acessar dados armazenados em arquivos de texto. O Python oferece a função open() para abrir e ler esses arquivos de maneira simples e eficiente.

Exemplo de como ler um arquivo .txt:
```python
# Abre o arquivo em modo leitura
with open('dados.txt', 'r') as file:
    conteudo = file.read()

print(conteudo)
```
- open('dados.txt', 'r'): Abre o arquivo dados.txt em modo de leitura.

- file.read(): Lê todo o conteúdo do arquivo e o armazena em conteudo.

- print(conteudo): Exibe o conteúdo do arquivo.

### Leitura de Arquivo Linha a Linha
Para processar arquivos de texto grandes ou quando você precisa analisar o conteúdo linha por linha (por exemplo, em arquivos CSV ou logs), é mais eficiente ler o arquivo linha a linha em vez de carregar todo o seu conteúdo de uma vez na memória com file.read().

O Python facilita isso, pois o objeto file retornado pela função open() é um iterador que produz uma linha do arquivo a cada iteração.

Exemplo de Leitura Linha a Linha
Você pode fazer a leitura linha a linha utilizando um loop for:

Python

**Abre o arquivo em modo leitura**

```python
with open('dados.txt', 'r') as file:
    # Itera sobre o objeto file para ler uma linha por vez
    for num_linha, linha in enumerate(file, 1):
        #  ('\n') é preservado no final de cada linha,
        # por isso usamos .strip() para removê-la e espaços em branco desnecessários.
        print(f"Linha {num_linha}: {linha.strip()}")
```
