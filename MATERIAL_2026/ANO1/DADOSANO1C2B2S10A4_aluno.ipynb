## Base Teórica

### O que é roteirizar um algoritmo?
Roteirizar um algoritmo é **planejar o raciocínio lógico** antes de programar: definir o problema, separar em passos e prever a execução.
Ajuda a **visualizar o fluxo**, **evitar erros lógicos** e **economizar tempo** (menos retrabalho).

### As três etapas do planejamento lógico
1. **Esboço de lógica (linguagem natural):** descreva o que o algoritmo deve fazer, sem sintaxe.
2. **Organização de etapas (pseudocódigo/fluxograma):** transforme o esboço em passos com **sequência**, **decisão** e **repetição**.
3. **Simulação da execução:** teste mentalmente com valores de exemplo (linha a linha), prevendo resultados e correções antes do código.

### Estruturas básicas envolvidas
| Tipo de estrutura | Função principal                           | Exemplo (pseudo)                 |
|-------------------|---------------------------------------------|----------------------------------|
| Sequência         | Executar passos em ordem                     | `ler dado → processar → imprimir`|
| Decisão           | Escolher caminho conforme condição           | `se x <= 100 então baixo senão ...` |
| Repetição         | Repetir enquanto condição for verdadeira     | `enquanto i < tamanho faça ...`  |

> **Boa prática:** em laços `while`, use condição **segura** (`i < len(lista)`) e **atualize o índice** uma vez por iteração.
