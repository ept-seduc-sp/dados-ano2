### Etapa 1: Importação e Carga de Dados

Iniciamos importando as bibliotecas necessárias para leitura de arquivos e geração de gráficos.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando os dois arquivos CSV 
df_a = pd.read_csv('DADOSANO1C2B4S21A3_auxiliar1.csv')
df_b = pd.read_csv('DADOSANO1C2B4S21A3_auxiliar2.csv')
```

### Etapa 2: 

Nesta etapa, chamamos a função scatter duas vezes no mesmo bloco de código para que os dados apareçam no mesmo eixo.

```python
# Plotando os dados da Turma A
plt.scatter(df_a['horas_estudo'], df_a['nota'], label='Turma A')

# Plotando os dados da Turma B no mesmo gráfico
plt.scatter(df_b['horas_estudo'], df_b['nota'], label='Turma B')

plt.legend() # Exibe a legenda para identificar cada turma
plt.show()
```

### Etapa 3: Personalização e Contraste Visual

Para facilitar a comparação, ajustamos a estética dos pontos. 

```python
# Customizando para diferenciar os grupos
plt.scatter(df_a['horas_estudo'], df_a['nota'], color='blue', marker='o', label='Turma A')
plt.scatter(df_b['horas_estudo'], df_b['nota'], color='red', marker='s', label='Turma B')

plt.title('Comparativo: Horas de Estudo vs Nota Final')
plt.xlabel('Horas de Estudo')
plt.ylabel('Nota Final')
plt.grid(True) # Adiciona grade para facilitar leitura
plt.legend()

plt.show()
```
