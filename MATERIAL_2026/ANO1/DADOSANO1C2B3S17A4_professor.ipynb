## Base Teórica

O Pandas é a principal biblioteca do Python para análise de dados. Sua estrutura central é o DataFrame, que funciona como uma tabela Excel, organizada em linhas e colunas. 

Para trabalhar com dados reais, nem sempre recebemos tabelas prontas; muitas vezes precisamos criar DataFrames a partir de dicionários ou listas, ajustar nomes de colunas para facilitar a leitura e criar novas métricas (colunas calculadas) que nos ajudem a extrair inteligência dos dados brutos.

**Exemplos**

**Criar DataFrame (Dicionário)**

Transforma dados estruturados em uma tabela.

```python
import pandas as pd

dados = {'Produto': ['A', 'B'], 'Preço': [10, 20]}
df = pd.DataFrame(dados)
print(df)
```
**Ajustar Nomes de Colunas**

Renomeia colunas para facilitar a manipulação.

```Python
df = df.rename(columns={'Preço': 'Valor_Unitario'})
# Ou define todas de uma vez:
df.columns = ['Item', 'Preço_RS']
```
**Criar Colunas Calculadas**

Gera novas métricas a partir de dados existentes.

```Python
df['Imposto'] = df['Preço_RS'] * 0.1
df['Total'] = df['Preço_RS'] + df['Imposto']
print(df)
```

## Atividade do Projeto da Unidade EcoMercado Semana 17: Refinando a Margem do EcoMercado

**Expectativa de entrega**

O aluno deve demonstrar o uso encadeado ou sequencial dos métodos de limpeza antes da subtração.

**Código exemplo esperado:**

````python
import pandas as pd
df_produtos = pd.read_csv('DADOSANO1C2B3S17A4_base_de_dados/produtos.csv')

# Limpeza e conversão necessária para a coluna vlr_preco
df_produtos['vlr_preco'] = df_produtos['vlr_preco'].str.replace('R$ ', '', regex=False).str.replace(',', '.').astype(float)

# Cálculo do Lucro
df_produtos['Lucro'] = df_produtos['vlr_preco'] - df_produtos['vlr_custo']

df_produtos.to_csv('DADOSANO1C2B3S17A4_base_de_dados/produtos_lucro.csv', index=False, encoding='utf-8-sig')
````

**Análise pedagógica**

Esta atividade consolida o conceito de Pipeline de Dados. O aluno não apenas carrega e processa, mas agora gera uma saída (output) que será consumida em aulas futuras. A dificuldade técnica reside na conversão de tipos, o que reforça o pensamento computacional e a atenção a padrões de dados (ponto vs vírgula).
