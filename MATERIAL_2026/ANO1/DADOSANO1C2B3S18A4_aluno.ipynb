## Base Teórica

Na Ciência de Dados, raramente os dados chegam perfeitos. É comum encontrar dados duplicados (devido a falhas de exportação ou erros humanos) e dados ausentes (campos não preenchidos).

- Duplicados: Usamos o método .duplicated() para identificar e .drop_duplicates() para remover.

- Ausentes (NaN): Representam a falta de informação. Podemos excluí-los com .dropna() ou preenchê-los com .fillna(). O preenchimento pode ser feito com um valor padrão (ex: 0 ou "Desconhecido") ou com medidas estatísticas como a média ou mediana.

**Exemplos**

### Tratamento de Dados Duplicados

Identificar e remover registros que se repetem de forma idêntica ou parcial.

```python
import pandas as pd

# Criando um DataFrame com duplicatas
df = pd.DataFrame({
    'ID': [1, 2, 2, 3],
    'Nome': ['Ana', 'Beto', 'Beto', 'Caio']
})

# Identificar linhas duplicadas
print(df.duplicated())

# Remover duplicadas (mantém a primeira ocorrência)
df_limpo = df.drop_duplicates()
```
### Tratamento de Dados Ausentes (NaN)

Lidar com lacunas de informação para evitar erros em cálculos estatísticos.

- A. Remover Dados Ausentes

    Ideal quando a quantidade de dados nulos é pequena e não prejudica a amostra.

    ```Python
    # Remove qualquer linha que contenha pelo menos um valor NaN
    df_sem_nulos = df.dropna()

    # Remove apenas se a linha inteira estiver vazia
    df_limpo_total = df.dropna(how='all')
    ```
- B. Preencher Dados Ausentes

    Técnica de imputação para manter a integridade do volume de dados.

    ```Python
    # Preencher com um valor fixo/padrão
    df['Status'] = df['Status'].fillna('Desconhecido')

    # Preencher com medidas estatísticas (Ex: Média)
    media_idade = df['Idade'].mean()
    df['Idade'] = df['Idade'].fillna(media_idade)

    # Preencher com a Mediana (mais robusta a outliers)
    mediana_salario = df['Salario'].median()
    df['Salario'] = df['Salario'].fillna(mediana_salario)
    ```
