## Base Teórica

A importação de dados é o primeiro e um dos mais cruciais passos em um pipeline de Ciência de Dados. 
No ecossistema Python, a biblioteca Pandas é a ferramenta padrão para essa tarefa. 
Ela oferece funções otimizadas como **pd.read_csv()**, **pd.read_excel()** e **pd.read_json()**, capazes de transformar arquivos brutos em DataFrames (tabelas estruturadas). 

Após a carga, é fundamental realizar uma inspeção inicial:

**.head():** Exibe as primeiras 5 linhas, permitindo uma visualização rápida da estrutura.

**.info():** Mostra um resumo técnico, incluindo o número de registros, valores nulos e o tipo de dado de cada coluna (inteiro, float, objeto/texto). Identificar se um número foi importado como texto, por exemplo, é vital para evitar erros em cálculos futuros.
