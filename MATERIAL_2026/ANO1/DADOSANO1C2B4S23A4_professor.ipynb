## Base Teórica

O Seaborn é uma biblioteca construída sobre o Matplotlib que facilita a criação de gráficos estatísticos atraentes e informativos. Enquanto o Matplotlib exige muitas linhas de código para customizações complexas, o Seaborn resolve isso com funções diretas que já tratam cores e legendas automaticamente.

**Exemplo: Visualizando Distribuições (Histplot)**

Para entender como uma variável se espalha (por exemplo, as notas dos filmes), usamos o histplot. O parâmetro kde=True adiciona uma linha suave que ajuda a ver a "forma" dos dados.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Exemplo: Distribuição de idades em um grupo
idades = [15, 16, 15, 17, 18, 16, 15, 16, 17, 16]
sns.histplot(idades, kde=True, color="green")
plt.title("Distribuição de Idades")
plt.show()
```
**Exemplo: Relação com Regressão (Regplot)**

Para verificar se uma variável influencia a outra (ex: tempo de estudo vs. nota), o regplot desenha os pontos e uma linha de tendência (regressão).

```python
# Exemplo: Horas de estudo vs Desempenho
horas = [1, 2, 3, 4, 5]
notas = [4, 5, 7, 8, 9]
sns.regplot(x=horas, y=notas)
plt.title("Tendência de Melhora")
plt.show()
```

## Atividade do Projeto da Unidade CineData Semana 23: Estratégia e Sucesso no Entretenimento

**Expectativa de entrega**

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('DADOSANO1C2B4S23A4_base_de_dados/base_de_filmes.csv')
sns.set_style("whitegrid")

# Gráfico de Dispersão Multivariado
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='orcamento_milhoes', y='faturamento_milhoes', hue='genero', s=100)
plt.title("Análise CineData: Orçamento vs Faturamento por Gênero")
plt.xlabel("Orçamento (Milhões)")
plt.ylabel("Faturamento (Milhões)")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()

# Evolução Temporal (Conforme Prática da Aula)
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='ano_lancamento', y='nota_publico')
plt.title("Evolução da Nota Média do Público ao Longo dos Anos")
plt.show()
````
