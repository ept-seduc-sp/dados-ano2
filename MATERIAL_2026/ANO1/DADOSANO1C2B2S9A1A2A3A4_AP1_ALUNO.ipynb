### Base teórica

Um **laço aninhado** é uma estrutura de repetição onde um **laço está dentro de outro.**

Ele é usado quando precisamos **percorrer duas ou mais dimensões de dados** — por exemplo, uma lista de listas, uma tabela ou o cruzamento de duas coleções de informações.

#### A Lógica

* O laço externo controla o ciclo principal (como as linhas de uma tabela);

* O laço interno executa várias vezes para cada repetição do laço externo (como as colunas de cada linha).

#### A Sintaxe

```python
for item_externo in conjunto_externo:
    for item_interno in conjunto_interno:
        # código a ser executado em cada combinação
```

### Raciocínio passo a passo:

1. O Python entra no primeiro laço (for item_externo) e escolhe um elemento.

2. Para cada elemento externo, o segundo laço (for item_interno) percorre todos os elementos internos.

3. O bloco de código interno é executado várias vezes, até que o laço interno termine.

4. Em seguida, o python retorna ao laço externo e repete o processo para o próximo elemento.

#### Imagine um relógio:

* O **ponteiro das horas** (laço externo) avança devagar;

* O **ponteiro dos minutos** (laço interno) completa várias voltas antes que a hora mude.

Assim também acontece com os laços aninhados — para cada passo do laço de fora, o laço de dentro percorre todo o seu ciclo.

#### **Exemplo — Cruzando cidades e produtos**

````python
# Lista de cidades analisadas
cidades = ["SP", "RJ"]

# Lista de produtos vendidos
produtos = ["Notebook", "Mouse"]

# Para cada cidade (laço externo)
for cidade in cidades:
    # Para cada produto (laço interno)
    for produto in produtos:
        # Exibe a combinação cidade-produto
        print("Cidade:", cidade, "| Produto:", produto)
````

> **_Raciocínio por trás do código_**
> 
> * O primeiro laço (externo) percorre todas as cidades da lista.
>
> * Para cada cidade, o segundo laço (interno) percorre todos os produtos.
>
> * A cada repetição, o Python cria uma combinação de cidade e produto.

### Construindo padrões com laços aninhados

Os laços aninhados também podem ser usados para criar padrões visuais simples no console.
Esse tipo de exercício ajuda a entender como os laços se repetem e como controlar linhas e colunas de forma organizada.

A ideia é pensar como se estivéssemos desenhando uma pequena tabela:

* O laço externo representa as linhas;

* O laço interno representa as colunas.

Esse tipo de estrutura é comum quando queremos exibir informações de forma alinhada, simulando uma planilha simples.

#### **Exemplo: organizando vendas por cidade e produto**

Um analista quer visualizar quantos produtos diferentes são oferecidos em cada cidade.

Ele ainda não tem uma planilha pronta, então vai gerar uma estrutura de tabela no console usando laços aninhados.

````python
# Listas com as cidades e os produtos disponíveis
cidades = ["SP", "RJ", "MG"]
produtos = ["Notebook", "Mouse", "Teclado"]

# Cabeçalho da tabela
print("CIDADE | PRODUTOS DISPONÍVEIS")
print("------------------------------")

# O laço externo percorre as cidades (linhas)
for cidade in cidades:
    # O laço interno percorre os produtos (colunas)
    for produto in produtos:
        print(cidade, "-", produto)
    print()  # pula uma linha entre as cidades
````

> **_Raciocínio:_**
> 
> O laço externo (for cidade in cidades) percorre as cidades, linha por linha.
>
> O laço interno (for produto in produtos) percorre os produtos e os associa à cidade atual.
>
> O resultado é uma visualização estruturada de dados, como se fosse uma mini-tabela.
