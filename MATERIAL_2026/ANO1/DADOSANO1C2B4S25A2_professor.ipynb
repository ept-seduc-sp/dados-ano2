## Gabarito — Cenário 1: Conhecendo os Dados 

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('DADOSANO1C2B4S25A2A3A4_base_de_dados/winequality-red.csv', sep=';')
print(df.head())
df.info()
````

## Gabarito — Cenário 2: Heatmap de Qualidade

````python
import matplotlib.pyplot as plt

# Selecionando colunas e calculando correlação
colunas = ['alcohol', 'pH', 'density', 'quality']
matriz_conhecimento = df[colunas].corr()

sns.heatmap(matriz_conhecimento, annot=True, cmap='YlGnBu')
plt.title('Matriz de Correlação - Componentes do Vinho')
plt.show()
````
