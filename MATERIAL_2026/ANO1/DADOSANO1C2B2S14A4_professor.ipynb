## Base Teórica

No dia a dia de um Cientista de Dados, é comum receber várias pequenas coletas e registros em formato .txt. Podem ser listas de clientes, nomes cadastrados em formulários, logs de acesso ou pequenos relatórios diários. Antes de qualquer análise, é fundamental saber como abrir esses arquivos, ler o conteúdo e, quando necessário, combinar dados de mais de uma fonte.
Python facilita bastante esse processo com a função open(), permitindo trabalhar com arquivos simples de forma rápida e organizada

**Exemplo — Leitura de arquivo simples**
```python
with open('dados.txt', 'r') as file:
    conteudo = file.read()

print(conteudo)


open('dados.txt','r'): abre o arquivo em modo leitura

read(): lê tudo como uma única string
```

**Exemplo — Lendo mais de um arquivo**
```python
with open('arquivo1.txt', 'r') as f1:
    a = f1.read()

with open('arquivo2.txt', 'r') as f2:
    b = f2.read()

print(a)
print(b)

```
- Cada arquivo é lido separadamente

- Depois os dados podem ser combinados
