## Base Teórica

Para que um gráfico seja eficiente, nem sempre a escala automática é a melhor escolha. Quando trabalhamos com Ciência de Dados, precisamos ajustar o "foco" da nossa imagem.

- Ajuste de Escala: Usamos o método plt.yscale('log'), por exemplo, quando temos valores muito pequenos próximos de valores muito grandes.

- Limites de Eixos: Os métodos plt.xlim(min, max) e plt.ylim(min, max) definem onde o gráfico começa e termina, evitando espaços vazios desnecessários ou focando em um detalhe.

- Personalização: Além dos eixos, o uso de plt.annotate() e cores ajuda a guiar o olhar do espectador para o que realmente importa.

Exemplo:


```python
import matplotlib.pyplot as plt

# Dados de exemplo
meses = ['Jan', 'Fev', 'Mar', 'Abr']
vendas = [10, 100, 1000, 10000]

plt.plot(meses, vendas, marker='o', color='purple')

# Alterando a escala para logarítmica para visualizar melhor o crescimento exponencial
plt.yscale('log')

# Definindo limites manuais para o eixo Y
plt.ylim(1, 20000)

plt.title("Exemplo de Escala Logarítmica")
plt.show()
```
