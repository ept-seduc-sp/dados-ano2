## Base Teórica

### Exemplo de Leitura de Arquivo

Em Ciência de Dados, frequentemente precisamos acessar dados armazenados em arquivos de texto. O Python oferece a função open() para abrir e ler esses arquivos de maneira simples e eficiente.

Exemplo de como ler um arquivo .txt:
```python
# Abre o arquivo em modo leitura
with open('dados.txt', 'r') as file:
    conteudo = file.read()

print(conteudo)
```
- open('dados.txt', 'r'): Abre o arquivo dados.txt em modo de leitura.

- file.read(): Lê todo o conteúdo do arquivo e o armazena em conteudo.

- print(conteudo): Exibe o conteúdo do arquivo.

### Exemplo de Uso do split()
O método split() é uma ferramenta útil para dividir uma string em uma lista de substrings, com base em um delimitador, como uma vírgula ou espaço. Isso é especialmente útil quando trabalhamos com dados separados por vírgulas ou outros caracteres em arquivos de texto.

Exemplo de como usar o split():
```python
# Dados de idades separados por vírgulas
idades = "25,30,22,28,35"

# Usa o split() para dividir as idades em uma lista
lista_idades = idades.split(',')

print(lista_idades)
```

- idades.split(','): Divide a string idades nas ocorrências de vírgulas e retorna uma lista com os valores de idade.
Resultado esperado:
['25', '30', '22', '28', '35']

## Cenário 2 — Criando um dicionário de produtos

**Expectativa de entrega:**
O aluno deve ser capaz de:

- Ler o arquivo e separar as idades com o método split().

- Usar list comprehension para converter as idades de string para inteiros.

- Calcular a soma das idades e a média corretamente.

**Código exemplo esperado:**

````python
with open('DADOSANO1C2B2S14A3_complementar1.txt', 'r') as file:
    conteudo = file.read()

# Separando as idades
idades = conteudo.split(',')

# Convertendo as idades para inteiros
lista_idades = [int(idade) for idade in idades]

# Calculando a soma das idades
soma_idades = sum(lista_idades)

# Calculando a média das idades
media_idades = soma_idades / len(lista_idades)

# Exibindo a média com duas casas decimais
print(f"A média das idades é: {media_idades:.2f}")
````
