## Base Teórica

**Exemplo de Código:**
```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Exemplo: Correlação entre tempo de estudo e nota
df = pd.DataFrame({'estudo': [1, 5, 10], 'nota': [4, 7, 10]}, 'tempo': [8,3,9])
dados_exemplo = df[['estudo', 'nota']] # selecionando apenas colunas específicas
correlacao = dados_exemplo.corr()

# Gerando o Heatmap
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.show()

# Gerando um Histplot de frequências
sns.histplot(dados_exemplo['nota'], bins=5)
plt.show()
```
