### Etapa 1: Carregamento

Importe o pandas e realize a leitura do arquivo de vendas para iniciar a análise.

```python
import pandas as pd
df = pd.read_csv('DADOSANO1C2B3S19A3_auxiliar1.csv')
df.head()
```

### Etapa 2: Criação do Agrupamento

Agrupe os dados pelas colunas de região, cidade e produto para criar uma visão segmentada.

```python
agrupado = df.groupby(['Regiao', 'Cidade', 'Produto'])
```

### Etapa 3: Cálculo Agregado

Aplique as funções de soma e contagem na coluna de faturamento para obter os totais e o volume de vendas.

```python
resultado = agrupado['Faturamento'].agg(['sum', 'count'])
```
