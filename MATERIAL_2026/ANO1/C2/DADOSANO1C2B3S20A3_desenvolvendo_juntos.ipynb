### Etapa 1: Carregar os dados e listar colunas

Antes de unir tabelas, precisamos carregar os arquivos e visualizar o nome exato das colunas para identificar a chave de ligação.

```python
import pandas as pd

df_leituras = pd.read_csv('DADOSANO1C2B3S20A3_auxiliar1.csv')
df_loc = pd.read_csv('DADOSANO1C2B3S20A3_auxiliar2.csv')

print("Colunas Leituras:", df_leituras.columns)
print("Colunas Localidades:", df_loc.columns)
```

### Etapa 2: Identificar a chave de ligação

Verifique se a coluna que identifica o sensor possui o mesmo nome em ambos os DataFrames (ex: cod_sensor). Se forem diferentes, precisaremos especificar no merge.

```python
# Verificando os tipos de dados da chave
print(df_leituras['cod_sensor'].dtype)
print(df_loc['cod_sensor'].dtype)
```

### Etapa 3: Executar o Merge

Utilizamos o pd.merge para realizar a junção (similar ao JOIN do SQL). O parâmetro on define a coluna em comum.

```python
# Quando how não é passado com parâmetro, o merge por padrão utiliza o inner.
df_unido = pd.merge(df_leituras, df_loc, on='cod_sensor') 
```
