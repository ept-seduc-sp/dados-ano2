### Etapa 1: Preparação e Dados

Vamos carregar o Seaborn e utilizar um dataset nativo ("tips") que contém dados sobre gorjetas em um restaurante.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando dados de exemplo
dados = sns.load_dataset("tips")
print(dados.head())
```

### Etapa 2: Comparando com hue

Vamos verificar se o valor da conta e da gorjeta varia conforme o cliente é fumante ou não.

```python
sns.scatterplot(data=dados, x="total_bill", y="tip", hue="smoker")
plt.title("Relação Conta x Gorjeta (Fumantes vs Não Fumantes)")
plt.show()
```

### Etapa 3: Criando Grades com FacetGrid

Agora adicionamos a linha que suaviza a visualização.

```python
# Criando a grade baseada no tempo (time)
grade = sns.FacetGrid(dados, col="time")

# Mapeando o gráfico de dispersão para a grade
grade.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```
