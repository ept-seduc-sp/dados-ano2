## Base Teórica

Para que um gráfico seja eficiente, nem sempre a escala automática é a melhor escolha. Quando trabalhamos com Ciência de Dados, precisamos ajustar o "foco" da nossa imagem.

- Ajuste de Escala: Usamos o método plt.yscale('log'), por exemplo, quando temos valores muito pequenos próximos de valores muito grandes.

- Limites de Eixos: Os métodos plt.xlim(min, max) e plt.ylim(min, max) definem onde o gráfico começa e termina, evitando espaços vazios desnecessários ou focando em um detalhe.

- Personalização: Além dos eixos, o uso de plt.annotate() e cores ajuda a guiar o olhar do espectador para o que realmente importa.

Exemplo:


```python
import matplotlib.pyplot as plt

# Dados de exemplo
meses = ['Jan', 'Fev', 'Mar', 'Abr']
vendas = [10, 100, 1000, 10000]

plt.plot(meses, vendas, marker='o', color='purple')

# Alterando a escala para logarítmica para visualizar melhor o crescimento exponencial
plt.yscale('log')

# Definindo limites manuais para o eixo Y
plt.ylim(1, 20000)

plt.title("Exemplo de Escala Logarítmica")
plt.show()
```

## Cenário 1 — Monitoramento de Crescimento de Micro-organismos

**Expectativa de Resposta:**

O uso da escala logarítmica é a resposta correta porque os dados crescem de forma geométrica. Sem ela, os valores iniciais (50, 200) ficam visualmente indistinguíveis no gráfico em comparação a 51.200.

````python
import matplotlib.pyplot as plt

tempo = [0, 1, 2, 3, 4, 5]
bacterias = [50, 200, 800, 3200, 12800, 51200]
plt.plot(tempo, bacterias, marker='s')
plt.yscale('log')
plt.title("Crescimento Bacteriano (Escala Log)")
plt.xlabel("Tempo (h)")
plt.ylabel("Qtd (Log)")
plt.show()
````

## Cenário 2 — Foco na Meta de Vendas da Cantina

**Expectativa de Resposta:**

O ajuste de ylim permite focar na meta. A anotação exercita a sensibilidade visual e criatividade do aluno.

````python
import matplotlib.pyplot as plt

dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex']
vendas = [12, 25, 30, 18, 10]
plt.bar(dias, vendas, color='orange')
plt.ylim(15, 35)
plt.annotate('Pico!', xy=('Qua', 30), xytext=('Qua', 31), ha='center')
plt.title("Foco em Vendas Acima da Meta")
plt.show()
````
