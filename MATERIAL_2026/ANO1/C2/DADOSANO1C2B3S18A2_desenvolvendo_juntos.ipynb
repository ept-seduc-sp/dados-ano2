```python
import pandas as pd
```

### Etapa 1: Ver nulos

Carregue o arquivo e veja onde faltam dados.

```python
df = pd.read_csv('DADOSANO1C2B3S18A2_auxiliar1.csv')
df.isnull().sum()
```

### Etapa 2: Remover críticos

Apague linhas onde o 'Produto' está vazio.

```python
df = df.dropna(subset=['Produto'])
```

### Etapa 3:  Preencher com Zero

Colunas de 'Desconto' vazias viram $0$.


```python
df['Desconto'] = df['Desconto'].fillna(0)
```

### Etapa 4:  Preencher com Média

Complete os preços vazios com a média da loja.

```python
df['Preco'] = df['Preco'].fillna(df['Preco'].mean())
```
