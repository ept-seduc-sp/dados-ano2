## Base Teórica

A biblioteca Matplotlib permite que a gente transforme números em imagens. Mas, para um gráfico ser bom, ele precisa ser "legível".

Exmeplos:

**1** Cores, Marcadores e Estilos de Linha
Você pode definir a estética da linha diretamente na função plt.plot().
.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Personalização: linha vermelha, tracejada e com marcadores circulares
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.show()
```
**2** Títulos e Rótulos (Labels)
Essenciais para dar contexto aos dados apresentados.

```python
plt.plot(x, y)

plt.title("Crescimento de Vendas") # Título principal
plt.xlabel("Meses")                # Eixo X
plt.ylabel("Receita (R$)")         # Eixo Y

plt.show()
```
**Exemplo Completo**
Unindo todos os conceitos para criar um gráfico legível:

```python
import matplotlib.pyplot as plt

# Dados
dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex']
temperatura = [22, 25, 21, 28, 24]

# Criando o gráfico com personalização combinada
plt.plot(dias, temperatura, color='magenta', marker='s', linestyle='-')

# Adicionando textos explicativos
plt.title("Variação de Temperatura na Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Graus Celsius (°C)")

# Exibe o gráfico
plt.show()
```

## Atividade do Projeto da Unidade CineData Semana 21: Primeiro Take — Mapeando o Mercado Cinematográfico

**Expectativa de entrega**

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('DADOSANO1C2B4S21A4_base_de_dados/base_de_filmes.csv')
contagem_generos = df['genero'].value_counts()

plt.figure(figsize=(10, 6))
contagem_generos.plot(kind='bar', color='darkorange')

plt.title('Distribuição de Filmes por Gênero - CineData')
plt.xlabel('Gênero')
plt.ylabel('Quantidade de Produções')
plt.xticks(rotation=45)
plt.show()
````

**Análise pedagógica**

Esta atividade inicia o projeto CineData conectando o objeto de conhecimento "Personalização Visual" a um problema real (entender o catálogo de filmes). Ao rotacionar os rótulos e adicionar títulos, o aluno demonstra a competência de comunicar dados com clareza. A análise mobiliza a curiosidade investigativa ao permitir que o aluno visualize, pela primeira vez, qual o gênero predominante na base de dados (ex: Drama ou Ficção Científica), preparando-o para as análises financeiras que virão nas próximas semanas
