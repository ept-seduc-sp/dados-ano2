## Base Teórica

Na ciência de dados, raramente as informações estão em uma só tabela. Usamos o merge() para unir arquivos diferentes através de uma coluna comum (chave).

- O que é: Funciona como um PROCV (Excel) ou JOIN (SQL).

- A lógica: Une linhas de tabelas distintas baseando-se em um código identificador (ex: id_prod).

**O Tipo: how='left'**

Garante que todos os dados da sua tabela principal (esquerda) permaneçam. Se não houver correspondência na segunda tabela, o Pandas preenche com NaN.

**Exemplos**

**1.** Vendas e Produtos (Adicionar nome ao ID).

```python
vendas = pd.DataFrame({'id': [1, 2], 'qtd': [10, 5]})
prods = pd.DataFrame({'id': [1], 'nome': ['Arroz']})

# O ID 2 permanece, mesmo sem nome cadastrado
print(vendas.merge(prods, on='id', how='left'))
```
**2.** Funcionários e Cargos (Relacionar nomes)

```python
func = pd.DataFrame({'nome': ['Ana', 'Beto']})
cargo = pd.DataFrame({'nome': ['Ana'], 'cargo': ['CEO']})

# Beto não é excluído da lista, seu cargo será NaN
print(func.merge(cargo, on='nome', how='left'))
```
