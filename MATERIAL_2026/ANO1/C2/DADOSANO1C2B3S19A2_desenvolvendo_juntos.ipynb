```python
import pandas as pd
```

### Etapa 1: Preparação

Carregue o dataset.

```python
df = pd.read_csv('DADOSANO1C2B3S19A2_auxiliar1.csv')
```

### Etapa 2: Exploração

Confira os tipos de dados das colunas.

```python
df.info()
```

### Etapa 3: Agrupamento Básico

Volume total de vendas por categoria.


```python
df.groupby('Categoria')['Vendas'].sum()
```

### Etapa 4: Uso do Agg

Cálculo de soma, média e frequência por grupo.

```python
df.groupby('Categoria')['Vendas'].agg(['sum', 'mean', 'count'])
```

### Etapa 5:  Análise

O professor explicará como ler essa tabela final: a soma mostra o volume total, a média indica o ticket médio por venda e o count mostra a frequência de pedidos.
