### Etapa 1: Criação/Carga dos Dados

Vamos ler os arquivos CSV que contêm as vendas de Janeiro e Fevereiro.

```python
df_jan = pd.read_csv('DADOSANO1C2B3S20A2_auxiliar1.csv')
df_fev = pd.read_csv('DADOSANO1C2B3S20A2_auxiliar2.csv')
print(df_jan)
print(df_fev)
```

### Etapa 2: Concatenação Simples (O Problema)

Ao unir os dados diretamente, o Pandas mantém os índices originais, o que gera duplicidade (ex: dois itens com índice 0).

```python
# Por padrão, o axis é 0 (concatenação vertical/linha abaixo de linha)
# Como não usamos ignore_index=True aqui, ele mantém os índices originais (0, 1, 2...) de cada tabela
df_errado = pd.concat([df_jan, df_fev])
print(df_errado) # Note que os índices se repetem
```

### Etapa 3: Correção de Índice

Utilizamos o parâmetro ignore_index=True para que o Pandas crie uma nova sequência numérica de 0 até o final da lista unida.


```python
df_final = pd.concat([df_jan, df_fev], ignore_index=True)
print(df_final)
```
