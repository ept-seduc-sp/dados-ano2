### Etapa 1: Preparação do Ambiente

Vamos importar a biblioteca e criar uma lista de vendas semanais.

```python
import matplotlib.pyplot as plt

semanas = [1, 2, 3, 4, 5]
vendas = [100, 120, 450, 130, 150] # Note o pico na semana 3
print("Ambiente pronto!")
```

### Etapa 2: Criando a Anotação de Pico

Agora, vamos identificar o valor "450" e apontar uma seta para ele explicando o motivo (ex: Black Friday).

```python
plt.figure(figsize=(8,4))
plt.plot(semanas, vendas, marker='o', color='gray')

# Adicionando a anotação
plt.annotate('Promoção Black Friday', 
             xy=(3, 450),      # Ponto do dado
             xytext=(3.5, 400), # Ponto do texto
             arrowprops=dict(facecolor='red', shrink=0.05))

plt.title('Vendas Mensais')
plt.show()
```
