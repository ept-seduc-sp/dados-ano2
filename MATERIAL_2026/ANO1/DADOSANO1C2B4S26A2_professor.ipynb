2. Tratamento de Dados Faltantes
Dados faltantes podem causar erros em cálculos. Podemos identificá-los com df.isnull().sum(). Se uma coluna importante tiver poucos valores nulos, podemos preenchê-los ou, em casos específicos, remover as linhas.

3. Transformação e Visualização
Transformar variáveis significa ajustar os dados para um formato melhor. Após a limpeza, usamos gráficos (como os de barras) para contar quantos itens existem em cada categoria.

```python
# Exemplo de contagem e gráfico

```
