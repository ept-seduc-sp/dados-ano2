### Limpeza de Dados com Conjuntos (Sets)
Imagine que você trabalha no marketing de uma empresa e recebeu uma lista gigante de e-mails de clientes. O problema? Muitos e-mails estão repetidos porque as pessoas se cadastraram em várias campanhas diferentes. Enviar mensagens repetidas ou gerar relatórios com esses erros é perda de tempo e passa uma imagem amadora.

Em Ciência de Dados, chamamos isso de duplicidade, e resolvê-la é parte essencial da Limpeza de Dados (Data Cleaning).

O que é um set?
O Python tem uma estrutura chamada set (conjunto). A principal característica dela é: ela não aceita valores repetidos. Se você tentar colocar o mesmo nome dez vezes, o set guardará apenas um.

**Exemplo 1: Criando um conjunto do zero**

**Conjuntos usam chaves {}**

```Python
cidades = {"São Paulo", "Rio de Janeiro", "Curitiba", "São Paulo"}
print(cidades) 
# Resultado: {'São Paulo', 'Rio de Janeiro', 'Curitiba'}
```
Como remover duplicatas de listas
Na vida real, os dados costumam vir em listas. Para "limpar" essa lista, usamos a função set() para transformar a lista em um conjunto e remover o que está sobrando automaticamente.

**Exemplo 2: O caso dos e-mails**

```Python
emails = ["ana@email.com", "joao@email.com", "ana@email.com", "bia@email.com"]

# A mágica acontece aqui:
unicos = set(emails)

print(unicos)
# Resultado no console: {'ana@email.com', 'joao@email.com', 'bia@email.com'}
```

Use o set() sempre que precisar de valores únicos. É a ferramenta mais rápida e simples para garantir que seus dados não tenham "lixo" ou repetições desnecessárias antes de começar qualquer análise séria.

### Operações com Conjuntos: União e Interseção
Além de remover duplicatas, os conjuntos permitem cruzar informações de bases de dados diferentes de forma muito simples. Imagine que você tem uma lista de clientes que compraram um produto e outra de clientes que clicaram num anúncio.

Exemplo 3: União e Interseção

```Python

compraram = {"Ana", "Carlos", "João"}
clicaram = {"Carlos", "Beatriz", "João"}

# União (Geralzão): Todos que interagiram, sem repetir ninguém
print("União:", compraram | clicaram)
# Resultado: {'Ana', 'Beatriz', 'Carlos', 'João'}

# Interseção (VIP): Apenas quem fez as duas coisas ao mesmo tempo
print("Interseção:", compraram & clicaram)
# Resultado: {'Carlos', 'João'}
```

**Entenda os símbolos:**
**|** (Barra vertical): Faz a União. Junta todos os nomes dos dois grupos e remove as repetições automaticamente.

**&** (E comercial): Faz a Interseção. Mostra apenas quem aparece nas duas listas simultaneamente.

Essas operações são fundamentais para entender o comportamento do público e focar as estratégias de marketing em grupos específicos.
