### Etapa 1: Preparação do Ambiente e Dados

Antes de desenhar, precisamos importar a biblioteca e definir nossos dados (Horário vs. Temperatura em °C).

```python
import matplotlib.pyplot as plt

# Dados de exemplo: Horas do dia e Temperatura da CPU
horas = ['08:00', '10:00', '12:00', '14:00', '16:00']
temp = [35, 42, 58, 65, 48]

print("Dados carregados com sucesso!")
```

### Etapa 2: Gráfico de Linha (Tendência)

O gráfico de linha é ideal para mostrar a variação da temperatura ao longo do tempo. 

```python
# Criando o gráfico de linha
plt.plot(horas, temp, color='blue')

plt.title('Variação de Temperatura da CPU')
plt.xlabel('Horário da Medição')
plt.ylabel('Temperatura (°C)')

plt.show()
```

### Etapa 3: Gráfico de Barra (Comparação)

Se quisermos comparar o impacto do calor em cada horário de forma isolada, o gráfico de barras é mais eficiente.

```python
# Criando o gráfico de barras
plt.bar(horas, temp, color='orange')

plt.title('Comparativo de Calor por Horário')
plt.xlabel('Horário')
plt.ylabel('Temperatura (°C)')

plt.show()
```
