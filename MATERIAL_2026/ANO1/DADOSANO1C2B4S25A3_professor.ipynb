## Base Teórica

**Exemplo de código:**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criando um exemplo simples (Pandas)
dados = pd.DataFrame({'X': [1, 2, 3], 'Y': [4, 5, 6], 'Categoria': ['A', 'B', 'A']})

# Configurando o layout (Matplotlib)
plt.figure(figsize=(6, 4))

# Criando o gráfico de dispersão com cores por categoria (Seaborn)
sns.scatterplot(data=dados, x='X', y='Y', hue='Categoria')

# Ajustando título e exibição (Matplotlib)
plt.title("Exemplo de Dispersão")
plt.show()
```

## Atividade Prática da Aula: Analisando as Pétalas da Íris

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregamento
df_iris = pd.read_csv('DADOSANO1C2B4S25A2A3A4_base_de_dados/Iris.csv')

# Configuração e Visualização
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title("Análise de Dispersão: Comprimento vs Largura da Sépala")
plt.xlabel("Comprimento da Sépala (cm)")
plt.ylabel("Largura da Sépala (cm)")
plt.show()
````
