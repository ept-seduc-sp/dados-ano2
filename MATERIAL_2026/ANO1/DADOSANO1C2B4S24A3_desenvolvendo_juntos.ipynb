### Etapa 1: Preparação e Tema Global

Primeiro, vamos importar as bibliotecas e definir um tema profissional de fundo claro com grades.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Definindo o estilo para 'whitegrid'
sns.set_theme(style="whitegrid")
print("Ambiente configurado!")
```

### Etapa 2: Criando os Dados e o Gráfico de Barras

Vamos simular o desempenho de quatro produtos.

```python
produtos = ['Smartphone', 'Tablet', 'Laptop', 'Fone']
vendas = [120, 85, 150, 200]

# Criando o gráfico e aplicando uma paleta de cores
sns.boxplot(x=produtos, y=vendas, hue=produtos, palette="magma") # Aqui o hue diz ao Seaborn que a cor deve variar de acordo com o nome do produto.
```

### Etapa 3: Finalização com Matplotlib

Agora vamos dar nome aos eixos e ao gráfico para que qualquer pessoa consiga entendê-lo.

```python
sns.boxplot(x=produtos, y=vendas, hue=produtos, palette="magma")
plt.title("Quantidade de Vendas por Produto")
plt.xlabel("Categoria de Produto")
plt.ylabel("Unidades Vendidas")

plt.show()
```
