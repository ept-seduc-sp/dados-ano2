**Exemplo de Código:**
```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um pequeno DataFrame de exemplo (Vendas de uma loja de games)
dados_vendas = {
    'Mês': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai'],
    'Jogos_Vendidos': [120, 150, 90, 200, 180],
    'Lucro_Milhares': [15, 22, 10, 35, 28]
}
df_games = pd.DataFrame(dados_vendas)

# 1. Criar a "Figura" e os "Eixos". 
# Queremos 2 linhas e 1 coluna (dois gráficos, um em cima do outro).
fig, eixos = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# 2. Criar o primeiro gráfico apontando para o eixo 0 (o de cima)
sns.barplot(x='Mês', y='Jogos_Vendidos', data=df_games, ax=eixos[0], color='skyblue')
eixos[0].set_title('Quantidade de Jogos Vendidos por Mês') # Título específico

# 3. Criar o segundo gráfico apontando para o eixo 1 (o de baixo)
sns.lineplot(x='Mês', y='Lucro_Milhares', data=df_games, ax=eixos[1], marker='o', color='green')
eixos[1].set_title('Evolução do Lucro (em Milhares de R$)') # Título específico

# 4. Ajustes finais
fig.suptitle('Painel de Desempenho - Loja de Games', fontsize=16) # Título GERAL da Imagem
plt.tight_layout() # Organiza os espaços para nada ficar sobreposto
plt.show()
```

## Atividade do Projeto da Unidade CineData Semana 26: Painel de Decisão Executiva

````python
# Passo 1: Importação
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Passo 2: Carga de Dados
df_filmes = pd.read_csv('DADOSANO1C2B4S26A2A3A4_base_de_dados/base_de_filmes.csv')

# Passo 3: Preparação da Tela (3 gráficos verticais)
fig, eixos = plt.subplots(nrows=3, ncols=1, figsize=(10, 15))

# Passo 4: Gráfico 1 (no primeiro eixo) - Faturamento por Gênero
sns.barplot(x='genero', y='faturamento_milhoes', data=df_filmes, ax=eixos[0], errorbar=None, palette='viridis')
eixos[0].set_title('Faturamento Médio por Gênero', fontsize=12)
eixos[0].set_xlabel('Gênero')
eixos[0].set_ylabel('Faturamento (Milhões)')

# Passo 5: Gráfico 2 (no segundo eixo) - Orçamento vs. Faturamento
sns.scatterplot(x='orcamento_milhoes', y='faturamento_milhoes', hue='genero', data=df_filmes, ax=eixos[1], alpha=0.7)
eixos[1].set_title('Relação entre Orçamento e Faturamento', fontsize=12)
eixos[1].set_xlabel('Orçamento (Milhões)')
eixos[1].set_ylabel('Faturamento (Milhões)')

# Passo 6: Gráfico 3 (no terceiro eixo) - Contagem de filmes por ano
sns.histplot(x='ano_lancamento', data=df_filmes, bins=20, ax=eixos[2], color='coral')
eixos[2].set_title('Evolução do Número de Lançamentos por Ano', fontsize=12)
eixos[2].set_xlabel('Ano de Lançamento')
eixos[2].set_ylabel('Quantidade de Filmes')

# Passo 7 e 8: Toque Profissional, Organização e Exibição
fig.suptitle('Painel de Análise Estratégica - CineData', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
````
