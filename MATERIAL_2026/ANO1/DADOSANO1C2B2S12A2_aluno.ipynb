## Base Teórica

### Métodos de lista

append(valor): adiciona no final.

insert(pos, valor): insere na posição indicada.

remove(valor): remove a 1ª ocorrência do valor.
````python
itens = ["A", "B"]
itens.append("C")     # ["A","B","C"]
itens.insert(1, "X")  # ["A","X","B","C"]
itens.remove("B")     # ["A","X","C"]
````

### Listas aninhadas

É uma “lista de listas”. Pense como uma tabela (linhas e colunas).
Você acessa assim: tabela[linha][coluna].

Exemplo: cada aluno tem [nome, nota].

```python
alunos = [["Ana", 9], ["João", 7]]

print(alunos[0])        # ["Ana", 9]  -> primeira linha
print(alunos[0][0])     # "Ana"       -> linha 0, coluna 0
print(alunos[0][1])     # 9           -> linha 0, coluna 1
print(alunos[0][0], alunos[0][1])  # Ana 9

```
Dica rápida para entender:

alunos[0] pega a primeira “linha”

alunos[0][1] pega o segundo item daquela linha

### List Comprehension

É um jeito mais curto e rápido de criar uma nova lista a partir de outra.
Você pode filtrar (pegar só alguns) e/ou transformar (modificar os valores).

Molde:
[expressao for item in lista if condicao]

- for item in lista percorre os elementos

- if condicao filtra (opcional)

- expressao define o que vai entrar na nova lista

Exemplo filtrando aprovadas (nota ≥ 7):

```python
notas = [5, 7, 9]
aprovadas = [n for n in notas if n >= 7]
print(aprovadas)  # [7, 9]

```

Exemplo filtrando e transformando ao mesmo tempo (dobrar só as notas ≥ 7):

```python
notas = [5, 7, 9]
dobro_aprovadas = [n * 2 for n in notas if n >= 7]
print(dobro_aprovadas)  # [14, 18]
```
