## Base Teórica

Em Ciência de Dados, é comum trabalhar com dados brutos e desorganizados vindos de diferentes fontes.
Essas informações precisam ser organizadas e agrupadas antes de gerar qualquer tipo de análise.

Os dicionários são úteis para armazenar pares de chave e valor (como nome e nota, produto e preço), enquanto os conjuntos (sets) permitem eliminar duplicatas e identificar valores únicos.

Também é possível converter estruturas, como transformar uma lista em um conjunto ou vice-versa, de acordo com o tipo de tratamento desejado.

Para criar um dicionário, use {} com pares chave: valor:
```python
produtos = {"mouse": 50, "teclado": 80}
```

Para exibir os dados:
```python
for nome, preco in produtos.items(): 
    print(nome, preco)
```
Para remover duplicatas, use set():
```python
valores_unicos = set(lista)
```

## Atividade 1: Organizando informações de clientes

**Expectativa de Resposta:**
O estudante deve criar um dicionário contendo tuplas para representar clientes, acessar as informações com .items() e usar o método .update() para adicionar novos dados.

**Código exemplo esperado:**

````python
# Dicionário com tuplas de dados
clientes = {
    "Ana": ("25", "São Paulo"),
    "Bruno": ("30", "Belo Horizonte"),
    "Carla": ("27", "Curitiba")
}

# Exibir nome e cidade
for nome, dados in clientes.items():
    print(f"{nome} mora em {dados[1]}")

# Adicionar novo cliente
clientes.update({"João": ("32", "Rio de Janeiro")})
print("\nCliente adicionado:", clientes)
````

**Atividade 2: Analisando vendas por produto**

**Expectativa de entrega:**
O estudante deve usar uma lista de tuplas para armazenar os produtos, aplicar um laço for para calcular o subtotal (preço × quantidade) e somar o faturamento total.

**Código exemplo esperado:**

````python
# Lista de produtos armazenada em tuplas
vendas = [
    ("Notebook", 3500, 3),
    ("Mouse", 80, 10),
    ("Teclado", 150, 5)
]

total = 0
for produto, preco, qtd in vendas:
    subtotal = preco * qtd
    total += subtotal
    print(f"{produto}: R$ {subtotal}")

print(f"\nFaturamento total: R$ {total}")
````

**Análise pedagógica:**
O exercício consolida a aplicação prática de tuplas e laços de repetição para resolução de problemas de negócios.
O aluno aprende a automatizar cálculos e interpretar resultados competências essenciais no trabalho com dados transacionais e indicadores de desempenho (KPIs) em projetos de Ciência de Dados.
