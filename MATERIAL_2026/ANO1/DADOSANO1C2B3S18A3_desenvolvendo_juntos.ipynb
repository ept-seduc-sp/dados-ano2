## Prática Guiada Passo a Passo

### Etapa 1: Carregamento

Importe o pandas e realize a leitura do arquivo CSV para iniciar a análise.

```python
import pandas as pd
df = pd.read_csv('DADOSANO1C2B3S18A3_auxiliar1.csv')
df.head()
```

### Etapa 2: Diagnóstico

Verifique a saúde dos dados identificando a quantidade de pedidos duplicados para reforçar a ideia de unicidade.

```python
df.duplicated(subset=['Pedido_ID']).sum()
```

### Etapa 3: Filtro de Subset

Remova as duplicatas focando apenas na coluna 'Pedido_ID' para garantir que cada pedido seja único.

```python
df_limpo = df.drop_duplicates(subset=['Pedido_ID'])
df_limpo.head()
```

### Etapa 4: Verificação

Compare o volume de dados e o resumo estrutural dos DataFrames para visualizar o impacto da limpeza.

```python
# Verificação de dimensões (impacto no número de linhas)
print(df.shape)
print(df_limpo.shape)

# Verificação estrutural inicial
df.info()

# Verificação estrutural final
df_limpo.info()
```
