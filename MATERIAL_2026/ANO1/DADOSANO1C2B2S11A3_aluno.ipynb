# Base Teórica

Programar em Python exige **organizar o código de forma lógica e estruturada**, usando **sequências, decisões e repetições**.  
Na **Ciência de Dados**, isso é essencial para **validar e limpar os dados**, garantindo resultados confiáveis.


### Estrutura do Código

Um código estruturado segue três blocos principais:

- **Sequência:** execução passo a passo das instruções.  
- **Decisão:** uso de `if`, `elif`, `else` para escolher caminhos.  
- **Repetição:** uso de `for` ou `while` para repetir tarefas.



### Controle de Laços

Durante a repetição, usamos comandos para controlar o fluxo:

- **`break`** → interrompe o laço (ex.: parar após muitos erros).  
- **`continue`** → pula para a próxima iteração (ex.: ignorar valores inválidos).

Esses comandos deixam o código mais eficiente e fácil de testar.
