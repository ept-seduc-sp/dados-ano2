## Base Teórica

Em programação, **laços de repetição** (`for` e `while`) permitem executar um conjunto de instruções várias vezes.  
No entanto, nem sempre queremos percorrer todas as etapas do loop.  

Para isso, usamos comandos especiais:

- **`break`**: **interrompe o laço** completamente quando uma condição é satisfeita.  
- **`continue`**: **pula apenas a repetição atual**, continuando o loop normalmente.  

Esses comandos ajudam a **organizar o código**, **evitar erros** e **economizar tempo de processamento**.

## Cenário 1 — Interrompendo um processo com `break`

Durante uma análise de dados de clientes, você precisa verificar se há **quantidade de items comprados inválidas (negativas)**.  
Assim que o programa encontrar o primeiro erro, ele deve **parar a execução**.

Atividade Prática
1. Observe o código abaixo e **rode-o no Jupyter Notebook**.

```python
quantidades = [20, 25, 30, -5, 28, 32]
i = 0
for quantidade in quantidades:
    if quantidade < 0:
        print(f"Erro encontrado na posição {i}: quantidade inválida ({quantidade})")
        break
    i+= 1
print("Verificação encerrada.")
```

## Cenário 2 — Pulando valores com `continue`

Você está calculando a média de notas de uma turma.  
Alguns alunos deixaram o campo vazio (valor `None`), e o programa deve **ignorar esses casos** sem interromper o cálculo.

### Atividade Prática
1. Observe o código abaixo e **execute-o no Jupyter Notebook**.  
2. Analise o comportamento do comando **`continue`** e **explique sua função**.  
3. Observe pelos *prints* se os valores `None` aparecem **antes da operação** `soma += nota`.


```python
notas = [8.0, None, 7.5, 6.0, None, 9.0]

soma = 0
contador = 0

for nota in notas:
    if nota is None: # Observe que se a nota é None, a condição if executa o continue
        continue
    print(nota)
    soma += nota
    contador += 1

media = soma / contador
print("Média final:", media)
```
