## Base Teórica

**Definição de tupla**
Tupla é uma coleção ordenada e imutável. Útil para registros fixos que não devem mudar por acidente (ex.: id, carimbo de data, UF).

```python
pedido = (1023, "2025-11-01", "MG")   # (id, data, UF)
print(type(pedido))   # <class 'tuple'>
```

**Diferenças entre listas e tuplas (essencial)**
- Mutabilidade: lista muda (append, remove, atribuição); tupla não muda.

- Símbolos: lista usa []; tupla usa ().

- Uso típico: lista para coleções que evoluem (carrinho de compras); tupla para registros/constantes (campos de um pedido).

- Desempenho: tupla tende a ser um pouco mais leve/rápida para leitura.

**Desempacotamento**

Atribui cada posição da tupla a uma variável — código fica mais legível:
(SKU é a sigla de Stock Keeping Unit — um código interno que identifica de forma única um item no estoque de uma empresa.)
```python
item = ("SKU123", "Teclado", 89.90)
sku, nome, preco = item
print(sku, nome, preco)  # SKU123 Teclado 89.9
```

Dica: sempre que um “registro” tiver campos estáveis (ex.: (id, data, UF)), prefira tupla; para coleções que crescem/alteram, prefira lista.
