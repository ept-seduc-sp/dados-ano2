### Base Te√≥rica

**Listas**

S√£o cole√ß√µes mut√°veis de elementos, criadas com colchetes [].
Permitem adicionar, remover e alterar valores ‚Äî ideais para dados que mudam com o tempo, como resultados de sensores ou listas de vendas.

```python

# Exemplo de lista

```

**Tuplas**

S√£o cole√ß√µes imut√°veis, criadas com par√™nteses ().
Usadas para armazenar dados fixos que n√£o devem ser alterados, como coordenadas geogr√°ficas, c√≥digos de produtos ou chaves de refer√™ncia.

```python
# Exemplo de tupla

print(produto[0])  # acesso por √≠ndice
# Sa√≠da: Notebook

```

| Caracter√≠stica | Lista           | Tupla              |
| -------------- | --------------- | ------------------ |
| Mutabilidade   | Pode alterar    |   N√£o pode alterar |
| Sintaxe        | `[]`            | `()`               |
| Uso comum      | Dados vari√°veis | Dados fixos        |
| Desempenho     | Mais lenta      | Mais r√°pida        |

### Exerc√≠cio 1: Cadastro e atualiza√ß√£o de dados de clientes

````python
# Lista de clientes (mut√°vel)
clientes = [
    ["Ana", "S√£o Paulo", 29],
    ["Carlos", "Campinas", 35],
    ["Lia", "Santos", 24]
]

# Atualizando dados do segundo cliente
clientes[1][0] = "Carlos Eduardo"
clientes[1][1] = "Rio de Janeiro"

# Adicionando novo cliente
clientes.append(["Jo√£o", "Belo Horizonte", 31])

# Exibindo lista atualizada
for c in clientes:
    print(f"Nome: {c[0]} | Cidade: {c[1]} | Idade: {c[2]}")
````
