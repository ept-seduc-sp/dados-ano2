## Base Teórica

O Pandas é a principal biblioteca do Python para análise de dados. Sua estrutura central é o DataFrame, que funciona como uma tabela Excel, organizada em linhas e colunas. 

Para trabalhar com dados reais, nem sempre recebemos tabelas prontas; muitas vezes precisamos criar DataFrames a partir de dicionários ou listas, ajustar nomes de colunas para facilitar a leitura e criar novas métricas (colunas calculadas) que nos ajudem a extrair inteligência dos dados brutos.

**Exemplos**

**Criar DataFrame (Dicionário)**

Transforma dados estruturados em uma tabela.

```python
import pandas as pd

dados = {'Produto': ['A', 'B'], 'Preço': [10, 20]}
df = pd.DataFrame(dados)
print(df)
```
**Ajustar Nomes de Colunas**

Renomeia colunas para facilitar a manipulação.

```Python
df = df.rename(columns={'Preço': 'Valor_Unitario'})
# Ou define todas de uma vez:
df.columns = ['Item', 'Preço_RS']
```
**Criar Colunas Calculadas**

Gera novas métricas a partir de dados existentes.

```Python
df['Imposto'] = df['Preço_RS'] * 0.1
df['Total'] = df['Preço_RS'] + df['Imposto']
print(df)
```
