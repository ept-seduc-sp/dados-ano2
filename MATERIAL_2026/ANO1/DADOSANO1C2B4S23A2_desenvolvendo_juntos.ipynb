### Etapa 1: Preparação do Ambiente

Vamos importar a biblioteca e criar uma lista de tempo de resposta de um suporte técnico (em minutos).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Simulando dados de tempo de resposta de um suporte técnico (em minutos)
tempos = [5, 7, 7, 8, 10, 12, 12, 12, 15, 18, 20, 25, 30]
df = pd.DataFrame(tempos, columns=['Tempo'])
print("Ambiente pronto!")
```

### Etapa 2: Criando o Histograma Base

O 'x' define a coluna dos dados.

```python
sns.histplot(data=df, x='Tempo', bins=6, color='skyblue')
plt.title('Distribuição de Tempo de Resposta')
plt.show()
```

### Etapa 3: Adicionando a Curva de Densidade (KDE)

Agora adicionamos a linha que suaviza a visualização.

```python
sns.histplot(data=df, x='Tempo', kde=True, color='purple')
plt.title('Distribuição com Curva de Densidade')
plt.show()
```
