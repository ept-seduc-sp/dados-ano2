## Base Teórica

Os dicionários em Python são estruturas que guardam pares de chave e valor.
Eles são amplamente usados em ciência de dados para representar informações estruturadas, como cadastros de clientes, produtos e relatórios.

### Dicionários Aninhados: A "Caixa dentro de Caixa"
Às vezes, uma informação é complexa e precisa de suas próprias etiquetas internas. Chamamos isso de Dicionários Aninhados. É como ter uma pasta para cada setor de uma empresa e, dentro dela, uma ficha com os detalhes.

Exemplo Prático:

 **Organizando os setores de uma empresa**

``` python
empresa = {
    "Vendas": {"funcionarios": 8, "meta": 50000},
    "TI": {"funcionarios": 5, "meta": 30000}
}
```

**Para acessar a meta de Vendas, usamos as duas "etiquetas":**
``` python
print(empresa["Vendas"]["meta"])  
# Resultado: 50000
```

**Acessando Valores do Dicionário** 

Para deixar seu código limpo e evitar erros (como tentar ler uma etiqueta que não existe), usamos três métodos principais:

- **.get() (O Buscador Seguro)**: Tenta ler uma etiqueta. Se ela não existir, ele te avisa em vez de travar o programa.

- **.update() (O Atualizador)**: Serve para trocar o valor de uma etiqueta ou colar uma etiqueta nova.

- **.pop() (O Removedor)**: Arranca uma etiqueta específica e remove a informação.

Exemplo Prático:

Python

- Cadastro de um cliente
```python
cliente = {"nome": "Marcos", "plano": "Premium"}
```

**Buscando um dado que pode não existir (telefone)**
1. Se não achar, exibe "Não cadastrado"
```python
print(cliente.get("telefone", "Não cadastrado"))
```
```python
2. Adicionando ou alterando dados
cliente.update({"telefone": "(11) 9999-8888"})
```
3. Removendo uma informação

```python
cliente.pop("plano")
print(cliente)
# Resultado final: {'nome': 'Marcos', 'telefone': '(11) 9999-8888'}
```

**Por que usar etiquetas?**

- Clareza: Em vez de dados[0], você usa dados["preco"]. Qualquer um entende o que aquilo significa.

- Organização: Você pode agrupar informações relacionadas em um só lugar.

- Velocidade: Encontrar um valor pela "etiqueta" é muito mais rápido do que percorrer uma lista inteira.

- a chave representa uma palavra

**Atividade Prática**

````python
frequencia_palavras = {
    "python": 3,
    "dados": 5,
    "ia": 2
}

print("Palavras analisadas:", list(frequencia_palavras.keys()))
print("Ocorrências da palavra 'dados':", frequencia_palavras.get("dados"))
print("Frequências registradas:", list(frequencia_palavras.values()))
````

## Atividade 2: Criando um dicionário de produtos

**Expectativa de entrega:**
O aluno deve criar um dicionário contendo, no mínimo, três produtos com preço e estoque, e exibir seus nomes e preços corretamente.

**Código exemplo esperado:**

````python
produtos = {
    "Mouse": {"preco": 50, "estoque": 120},
    "Teclado": {"preco": 80, "estoque": 75},
    "Monitor": {"preco": 600, "estoque": 30}
}

for nome, dados in produtos.items():
    print(f"{nome} - R$ {dados['preco']}")
````
