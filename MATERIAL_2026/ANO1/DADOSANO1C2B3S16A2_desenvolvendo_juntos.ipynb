### Etapa 1: Carregamento dos dados

Vamos ler o arquivo CSV base para ter um DataFrame pronto para exportar.

```python
df = pd.read_csv('DADOSANO1C2B3S16A2_auxiliar1.csv')
```

### Etapa 2: Exportação para Excel (Básico)

Criaremos o primeiro arquivo .xlsx.

```python
df.to_excel('relatorio_vendas.xlsx')
```

### Etapa 3: Ajuste de Índice
Notem como o Excel ganha uma coluna de números. Vamos removê-la para o relatório ficar limpo.


```python
df.to_excel('relatorio_limpo.xlsx', index=False)
```
