## Base Teórica

Em um projeto de Ciência de Dados, é comum precisar tomar decisões automáticas e repetir ações até que uma condição seja satisfeita.

Nessas situações, usamos estruturas condicionais (`if`) e laços de repetição (`while`) juntos, para controlar o fluxo do programa.

O `while` executa um conjunto de instruções enquanto uma condição for verdadeira.

Quando essa condição se torna falsa, o programa para automaticamente.

Essa combinação de decisão + repetição permite automatizar verificações, contar eventos e interromper processos sem intervenção humana.

**Exemplo 1 — Verificando qualidade dos dados**

Suponha que você tenha uma lista de valores numéricos representando registros de temperatura coletados por sensores.

Durante a leitura, é importante garantir que todos os valores estejam dentro de uma faixa aceitável — por exemplo, entre 10 °C e 90 °C.

Se o programa encontrar um valor fora dessa faixa, ele deve parar e avisar que há um possível erro no conjunto de dados.

```python
temperaturas = [35, 40, 44, 88, 95, 32, 30]

i = 0
alerta = False

while i < len(temperaturas) and not alerta:
    if temperaturas[i] < 10 or temperaturas[i] > 90:
        print("Valor fora do intervalo:", temperaturas[i])
        alerta = True
    i += 1

if not alerta:
    print("Todos os valores estão dentro do intervalo permitido.")
````

> **O que acontece no código:**
>
> * O laço percorre cada valor da lista.
>
> * A condição do if identifica valores fora da faixa.
>
> * Quando encontra um valor inválido, ativa o alerta e o laço é interrompido automaticamente.
>
> Esse tipo de verificação é muito usado para limpeza e validação de dados em projetos reais.

**Exemplo 2 — Contando valores válidos**

Outra aplicação frequente é contar quantos dados atendem a um critério específico.

Por exemplo, contar quantos registros são válidos (dentro da faixa esperada).

```python
temperaturas = [35, 40, 44, 88, 95, 32, 30]

i = 0
validos = 0

while i < len(temperaturas):
    if 10 <= temperaturas[i] <= 90:
        validos += 1
    i += 1

print("Total de registros válidos:", validos)
```

> **O que acontece no código:**
>
> - O `while` percorre toda a lista.
>
> - A condição 10 <= valor <= 90 identifica registros corretos.
>
> - O contador validos soma cada ocorrência.
>
> Esses dois exemplos mostram como combinar decisão e repetição para analisar dados automaticamente.

### Exercício 1 – Filtrando notas válidas

`````python
notas = [8, 9, 11, 7, -1, 10, 5, 12, 6]

i = 0
validas = 0

while i < len(notas):
    if 0 <= notas[i] <= 10:
        validas += 1
    i += 1

print("Total de notas válidas:", validas)
`````

> ***Explicação para o professor:***
>
> * O código percorre uma **lista de notas** e verifica se cada valor está **dentro do intervalo de 0 a 10**.  
> * A cada iteração, o laço `while` analisa um elemento e o `if` decide se ele é **válido**.  
> * O contador `validas` soma todas as notas corretas encontradas.  
> * Esse exercício reforça o uso conjunto de **decisão e repetição** para validar dados em uma lista.  
> * A estrutura é simples e ajuda o aluno a perceber a lógica de verificação automática.


### Exercício 2 - Interrompendo a leitura quando há erro

`````python
vendas = [120, 300, 250, -50, 400, 200]

i = 0
erro = False

while i < len(vendas) and not erro:
    if vendas[i] < 0:
        print("Erro: valor de venda inválido encontrado:", vendas[i])
        erro = True
    i += 1

if not erro:
    print("Todos os valores são válidos.")
`````

> ***Explicação para o professor:***
>
> * O código percorre uma **lista de vendas diárias** e procura um valor **negativo**, que indica erro.  
> * O laço `while` utiliza a condição `not erro_encontrado` para **controlar a parada** do programa.  
> * Ao encontrar o valor incorreto, a **flag booleana** é ativada e o laço encerra automaticamente.  
> * Essa lógica mostra como o `while` pode ser usado para **interromper leituras de dados** sem precisar de `break`.  
> * O exercício consolida o raciocínio de **controle de repetição com condição**.
