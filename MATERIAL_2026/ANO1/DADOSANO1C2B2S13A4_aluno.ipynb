## Base Teórica

Em Ciência de Dados, é comum trabalhar com dados brutos e desorganizados vindos de diferentes fontes.
Essas informações precisam ser organizadas e agrupadas antes de gerar qualquer tipo de análise.

Os dicionários são úteis para armazenar pares de chave e valor (como nome e nota, produto e preço), enquanto os conjuntos (sets) permitem eliminar duplicatas e identificar valores únicos.

Também é possível converter estruturas, como transformar uma lista em um conjunto ou vice-versa, de acordo com o tipo de tratamento desejado.

Para criar um dicionário, use {} com pares chave: valor:
```python
produtos = {"mouse": 50, "teclado": 80}
```

Para exibir os dados:
```python
for nome, preco in produtos.items(): 
    print(nome, preco)
```
Para remover duplicatas, use set():
```python
valores_unicos = set(lista)
```
