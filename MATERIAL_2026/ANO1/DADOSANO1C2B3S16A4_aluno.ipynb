## Base Teórica

Exportar dados é a etapa final de um processo de tratamento. Após limpar e organizar as informações em um DataFrame (usando a biblioteca Pandas), precisamos salvar esse resultado para que outras pessoas ou sistemas possam utilizá-lo.

 Os formatos mais comuns são o **CSV (leve e universal)** e o **Excel (ideal para relatórios administrativos)**. O comando principal que utilizamos é o **.to_csv()** ou **.to_excel()**, sempre lembrando de gerenciar o parâmetro index=False para evitar colunas desnecessárias de numeração.

**Exemplos de Exportação**

Ao exportar, o parâmetro index=False é fundamental para não criar uma coluna de índices automáticos no seu arquivo final.

```Python
import pandas as pd

# Criando um DataFrame de exemplo
df = pd.DataFrame({'Produto': ['A', 'B'], 'Preço': [10, 20]})

# 1. Exportar para CSV
df.to_csv('dados_finais.csv', index=False, sep=';', encoding='utf-8')

# 2. Exportar para Excel
df.to_excel('relatorio.xlsx', index=False, sheet_name='Vendas')
```
**Verificação de Arquivo**

```python
import os
```
