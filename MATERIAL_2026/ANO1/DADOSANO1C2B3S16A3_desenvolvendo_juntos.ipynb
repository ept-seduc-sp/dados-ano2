## Prática Guiada Passo a Passo

### Etapa 1: Importação de Dados

Começaremos criando um DataFrame simples para teste.

```python
import pandas as pd
import os

data = {
    'Produto': ['Notebook', 'Smartphone', 'Monitor'],
    'Preco': [4500, 2500, 1200]
}
df = pd.DataFrame(data)
print("DataFrame criado com sucesso.")
```

### Etapa 2: Simulando o Erro

Tentaremos salvar em um caminho que não existe para observar o erro do Python.

```python
# Isso resultará em erro se a pasta 'relatorios' não existir

df.to_csv('relatorios/dados_teste.csv', index=False)
```

### Etapa 3: Verificação de Existência
Usaremos os.path.exists() para checar se o arquivo já existe e perguntar ao "sistema" (ou exibir um aviso) antes de sobrescrever.

```python
caminho = 'dados_teste.csv'

if os.path.exists(caminho):
    print(f"Aviso: O arquivo '{caminho}' já existe. Deseja sobrescrever?")
else:
    df.to_csv(caminho, index=False)
    print("Arquivo salvo com sucesso.")
```

### Etapa 4: Automação de Pastas
Utilizaremos **os.makedirs()** para criar a pasta caso ela não exista, garantindo o sucesso da exportação.

```python
pasta = 'exportacao/final/'
if not os.path.exists(pasta):
    os.makedirs(pasta) # cria a pasta caso ela não exista
    print(f"Diretório '{pasta}' criado.")

df.to_csv(pasta + 'dados_finais.csv', index=False) #salva
print("Exportação concluída com sucesso.")
```
