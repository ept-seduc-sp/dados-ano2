## Base Teórica

Para entender o desempenho de um negócio, raramente olhamos para vendas individuais. Precisamos "dar um passo para trás" e observar o comportamento dos grupos. No Pandas, utilizamos o método `groupby()` para isso.

Imagine que você tem uma lista de vendas de frutas. O `groupby('Item')` junta todas as "Maçãs" em um bloco e todas as "Bananas" em outro. Após agrupar, aplicamos funções matemáticas para extrair insights:

* **`.sum()`**: Para saber o total vendido (Faturamento).
* **`.mean()`**: Para saber o valor médio de cada venda (Ticket Médio).
* **`.count()`**: Para saber quantas vendas foram realizadas.


### 1. O Método `.agg()` (Múltiplas Métricas)
Enquanto funções simples aplicam apenas um cálculo, o `.agg()` permite realizar **várias operações simultâneas** em colunas diferentes, gerando um resumo completo de uma só vez.

```python
# Criando um resumo estatístico por Setor
resumo = df.groupby('Setor').agg({
    'Preco': ['sum', 'mean'],  # Faturamento e Ticket Médio
    'Item': 'count'            # Quantidade total de vendas
})
```
### 2. O Método reset_index() (Organização)
Por padrão, o Pandas transforma a coluna agrupada (ex: 'Setor') no índice da tabela. O reset_index() restaura essa coluna para o corpo do DataFrame, facilitando a visualização e futuras filtragens.

```python
# Agrupando e transformando em um DataFrame comum (tabela plana)
total_por_setor = df.groupby('Setor')['Preco'].sum().reset_index()
```
