## Base Teórica

Na ciência de dados, nem sempre os eixos automáticos do Python são os melhores para contar uma história. Às vezes, precisamos "dar um zoom" em uma parte do gráfico ou mudar os nomes que aparecem nos eixos para que qualquer pessoa entenda.

**Exemplo: Customizando nomes no eixo X**

Se temos números (0, 1) representando categorias, podemos trocá-los por textos:


```python
import matplotlib.pyplot as plt

x = [0, 1]
y = [100, 250]

plt.bar(x, y)
# Customizando os "ticks" (marquinhas) do eixo X
plt.xticks(x, ['Grupo A', 'Grupo B']) 
plt.show()
```
**Exemplo: Definindo limites no eixo Y**

Para evitar que o gráfico pareça "achatado" ou para focar em um intervalo:

```python
plt.scatter(x, y)
plt.ylim(0, 500) # O eixo Y agora vai obrigatoriamente de 0 a 500
plt.show()
```

## Atividade do Projeto da Unidade CineData Semana 22: O Sucesso em Gráficos

**Expectativa de entrega**

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('DADOSANO1C2B4S22A4_base_de_dados/base_de_filmes.csv')

# Gráfico de Barras
faturamento_medio = df.groupby('genero')['faturamento_milhoes'].mean()
faturamento_medio.plot(kind='bar', color='skyblue')
plt.xticks(rotation=45)
plt.title("Faturamento Médio por Gênero")
plt.show()

# Gráfico de Dispersão
plt.scatter(df['orcamento_milhoes'], df['faturamento_milhoes'], alpha=0.5)
plt.xlabel("Orçamento (Milhões)")
plt.ylabel("Faturamento (Milhões)")

plt.annotate("Sucesso de Bilheteria", xy=(200, 1250), xytext=(150, 1000),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.title("Relação Orçamento vs Faturamento")
plt.show()
````

**Análise pedagógica**

O projeto articula os objetos macro (projetos gráficos) e micro (limites e anotações). O uso da rotação nos nomes dos gêneros resolve um problema comum de legibilidade, enquanto a anotação no gráfico de dispersão introduz a capacidade de destacar "insights" específicos, essencial para um Cientista de Dados Júnior.
