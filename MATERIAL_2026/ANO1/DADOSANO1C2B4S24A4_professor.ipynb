## Base Teórica

A estilização no Seaborn é feita principalmente através do método sns.set_theme(), que altera o fundo, as fontes e o estilo das linhas. Além disso, as paletas de cores (palette) permitem que os dados sejam diferenciados de forma harmônica. O parâmetro hue é essencial para segmentar os dados por categorias (como "gênero") dentro de um mesmo gráfico.

Exemplo de Estilização:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Definindo o tema e a paleta de cores global
sns.set_theme(style="darkgrid", palette="pastel")

# Exemplo de gráfico estilizado com hue
sns.scatterplot(data=df, x="coluna_x", y="coluna_y", hue="categoria")
plt.show()
```

**Atividade Prática**

1. Refinamento do Gráfico de Barras (Semana 22):
    - Carregue o arquivo base_de_filmes.csv.
    - Use sns.set_theme(style="whitegrid") para preparar o ambiente.
    - Altere o gráfico de faturamento médio por gênero utilizando o sns.barplot do Seaborn com a paleta palette="viridis".

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('DADOSANO1C2B4S24A4_base_de_dados/base_de_filmes.csv')

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,5))
faturamento_medio = df.groupby('genero')['faturamento_milhoes'].mean().reset_index()
sns.barplot(data=faturamento_medio, x='genero', y='faturamento_milhoes') # Adicione o pallette aqui
plt.xticks(rotation=45)
plt.title("Faturamento Médio por Gênero (Estilizado)")
plt.show()
````

2. Refinamento do Gráfico de Dispersão (Semana 22):

    - Veja o código do gráfico de dispersão (Scatter Plot).

    - O plt.scatter foi alterado para o sns.scatterplot do Seaborn.

    - Adicione o parâmetro hue="pais" para verificar se a origem do filme influencia na relação orçamento/faturamento.

````python
sns.set_theme(style="dark")
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='orcamento_milhoes', y='faturamento_milhoes',  palette="deep") # Adicione o hue aqui
plt.title("Relação Orçamento vs Faturamento por País")
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
````

3. Refinamento da Evolução Temporal (Semana 23):

    - O sns.lineplot foi utilizado para mostrar a nota do público ao longo dos anos.

    - Aplique uma paleta de cor específica (ex: color="red") e mude o estilo do tema para style="ticks".

````python
# Adicione o tema aqui
plt.figure(figsize=(10,4))
sns.lineplot(data=df, x='ano_lancamento', y='nota_publico') # Adicione o Color aqui
plt.title("Tendência de Notas ao Longo do Tempo")
plt.show()
````

**Observe como a aplicação de temas e paletas de cores transformou seus gráficos, tornando a análise muito mais profissional e estilizada em comparação às práticas anteriores!**

# Gabarito / Expectativas de resposta – Orientações para o professor

## Atividade do Projeto da Unidade CineData Semana 24: Finalização Visual da Análise de Sucesso

**Expectativa de entrega**

O aluno deve demonstrar a transição de gráficos básicos do Matplotlib para gráficos estilizados do Seaborn, utilizando temas globais e diferenciação por cores (hue).

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('DADOSANO1C2B4S24A4_base_de_dados/base_de_filmes.csv')

# 1. Gráfico de Barras Estilizado
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,5))
faturamento_medio = df.groupby('genero')['faturamento_milhoes'].mean().reset_index()
sns.barplot(data=faturamento_medio, x='genero', y='faturamento_milhoes', palette="viridis")
plt.xticks(rotation=45)
plt.title("Faturamento Médio por Gênero (Estilizado)")
plt.show()

# 2. Scatter Plot com hue e tema dark
sns.set_theme(style="dark")
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='orcamento_milhoes', y='faturamento_milhoes', hue='pais', palette="deep")
plt.title("Relação Orçamento vs Faturamento por País")
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()

# 3. Evolução Temporal com estilo ticks
sns.set_theme(style="ticks")
plt.figure(figsize=(10,4))
sns.lineplot(data=df, x='ano_lancamento', y='nota_publico', color="orange")
plt.title("Tendência de Notas ao Longo do Tempo")
plt.show()
````

**Análise pedagógica**

A atividade consolida o aprendizado da semana ao transformar códigos funcionais, mas visualmente simples (Matplotlib), em entregáveis profissionais (Seaborn). O uso de set_theme ensina ao aluno que a visualização de dados não é apenas sobre "fazer o gráfico", mas sobre garantir a legibilidade e a estética para o stakeholder (a produtora CineData). A aplicação do hue em diferentes contextos (gênero e país) reforça a análise multivariada.
