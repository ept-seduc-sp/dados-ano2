## Objetivo

Compreender como manipular datas em DataFrames no PySpark, realizando conversões, extrações e formatações.

Durante a atividade vamos utilizar as seguintes funções:

- `to_date()`
- `year()`
- `month()`
- `date_format()`

## Preparando o ambiente

Antes de trabalhar com DataFrames no PySpark, precisamos iniciar uma sessão Spark.

A `SparkSession` é o ponto de entrada para trabalhar com dados no PySpark.


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date, year, month, date_format

spark = SparkSession.builder \
    .appName("ManipulacaoDatas") \
    .getOrCreate()
```

## Leitura do arquivo de dados

Agora vamos carregar um conjunto de dados com informações de vendas.

Arquivo utilizado:

`DADOSANO2C6B3S20A1_auxiliar.csv`

O objetivo é trabalhar com a coluna de data para realizar transformações.

```python
df = spark.read.csv(
    "DADOSANO2C6B3S20A1_auxiliar.csv",
    header=True,
    inferSchema=True
)

df.show()
```

## Conversão de datas com `to_date()`

A função `to_date()` é utilizada para converter uma coluna de texto em formato de data.

Isso é importante porque operações com datas só funcionam corretamente quando o tipo da coluna é adequado.

```python
df = df.withColumn("data_convertida", to_date(col("data")))

df.show()
```

## Extraindo o ano com `year()`

A função `year()` permite extrair o ano de uma data.

Isso é útil para agrupar dados por ano ou analisar tendências ao longo do tempo.

```python
df = df.withColumn("ano", year(col("data_convertida")))

df.show()
```

## Extraindo o mês com `month()`

A função `month()` retorna o mês da data.

Essa informação é muito utilizada em análises como vendas por mês.

```python
df = df.withColumn("mes", month(col("data_convertida")))

df.show()
```

## Formatando datas com `date_format()`

A função `date_format()` permite alterar a forma como a data é exibida.

Isso é útil para relatórios e visualizações.

```python
df = df.withColumn(
    "data_formatada",
    date_format(col("data_convertida"), "dd/MM/yyyy")
)

df.show()
```
