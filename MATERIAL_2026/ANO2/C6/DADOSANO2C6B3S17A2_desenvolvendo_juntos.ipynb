## Objetivos
- Carregar arquivos CSV e JSON em DataFrames no PySpark.
- Visualizar dados com `show()` para uma inspeção inicial.
- Entender a estrutura de cada DataFrame com `printSchema()`.
- Gerar estatísticas descritivas básicas com `describe()`.
- Comparar rapidamente CSV e JSON quanto a estrutura e tipos de dados.


## Lista de Materiais

- Ambiente com PySpark configurado.
- Jupyter Notebook (este arquivo).
- Arquivos auxiliares no mesmo diretório:
  - `DADOSANO2C6B3S17A2_auxiliar.csv`
  - `DADOSANO2C6B3S17A2_auxiliar.json`


### Etapa 1: Iniciando a SparkSession
A **SparkSession** é o ponto de entrada do PySpark.
Nesta aula, ela será usada para ler os arquivos e criar os DataFrames.


```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("S17A2_Exploracao_Inicial")
    .master("local[*]")
    .getOrCreate()
)

print("SparkSession ativa")
```

### O que fizemos aqui?
- Importamos `SparkSession`.
- Inicializamos a sessão Spark local para execução no computador.
- Guardamos a sessão na variável `spark`, que será usada nas leituras.


### Etapa 2: Definindo os caminhos dos arquivos
Antes de ler os dados, organizamos os caminhos do CSV e do JSON em variáveis.
Isso facilita reaproveitar o código e evita repetir nomes de arquivos.


```python
csv_path = "DADOSANO2C6B3S17A2_auxiliar.csv"
json_path = "DADOSANO2C6B3S17A2_auxiliar.json"

csv_path, json_path
```

### O que observar?
- `csv_path` e `json_path` concentram os nomes dos arquivos em um único lugar.
- Se o arquivo mudar de nome ou pasta, basta ajustar essas variáveis.


### Etapa 3: Leitura do CSV
Agora vamos ler o CSV com duas opções importantes:

- `header = true`: usa a primeira linha como nome das colunas.
- `inferSchema = true`: tenta identificar tipos como número e texto.


```python
df_csv = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(csv_path)
)

print("CSV carregado")
```

### Resultado esperado desta etapa
Com essas opções, o DataFrame tende a ficar mais pronto para análise,
pois já chega com nomes de colunas e tipos mais adequados.


### Etapa 4: Leitura do JSON
Nesta etapa, carregamos o arquivo JSON para comparar com o CSV.


```python
df_json = spark.read.option("multiLine", "true").json(json_path)

print("JSON carregado")
```

### O que fizemos nesta etapa?
- Criamos `df_json` a partir do arquivo `.json`.
- A partir daqui, temos dois DataFrames para explorar: `df_csv` e `df_json`.


### Etapa 5: Visualização inicial com `show()`
A função `show()` ajuda a fazer uma leitura rápida dos registros.
Isso permite verificar se o conteúdo foi carregado corretamente.


```python
print("Amostra do CSV:")
df_csv.show(5, truncate=False)

print("\nAmostra do JSON:")
df_json.show(5, truncate=False)
```

### Interpretação rápida
- `show(5)` exibe as primeiras 5 linhas.
- `truncate=False` evita cortar textos e facilita a inspeção visual.
- Essa verificação inicial costuma ser o primeiro passo antes de transformar dados.


### Etapa 6: Estrutura com `printSchema()`
Agora vamos observar os tipos de cada coluna (ex.: `string`, `integer`, `double`).


```python
print("Schema do CSV:")
df_csv.printSchema()

print("\nSchema do JSON:")
df_json.printSchema()
```

### O que analisar no schema?
- Quais colunas aparecem como numéricas.
- Quais colunas aparecem como texto.
- Se há diferenças de tipos entre CSV e JSON.


### Etapa 7: Estatísticas com `describe()`
Por fim, usamos `describe()` para obter estatísticas descritivas básicas.


```python
print("Describe do CSV:")
df_csv.describe().show()

print("\nDescribe do JSON:")
df_json.describe().show()
```

### Como interpretar `describe()`
- O resultado mostra `count`, `mean`, `stddev`, `min` e `max`.
- Em colunas numéricas, esses valores ajudam a ter uma noção geral da distribuição.
- Em colunas textuais, o comportamento pode variar (nem sempre há média/desvio).

Esses passos apoiam decisões de limpeza, transformação e análise nas próximas etapas.
