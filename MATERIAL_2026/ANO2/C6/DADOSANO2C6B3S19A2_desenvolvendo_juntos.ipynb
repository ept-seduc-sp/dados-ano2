## Objetivo
Compreender como agrupar dados em um DataFrame com PySpark e aplicar agregações para gerar informações resumidas por categoria.

Durante a atividade vamos utilizar as seguintes funções e recursos:

• `groupBy()`  
• `count()`  
• `sum()`  
• `avg()`  
• `agg()`  

Essas operações são muito utilizadas na análise de dados, pois ajudam a transformar registros detalhados em métricas que apoiam a interpretação e a tomada de decisão.

## Preparando o ambiente

Antes de trabalhar com agrupamentos e agregações no PySpark, precisamos iniciar uma sessão Spark.

A `SparkSession` é o ponto de entrada para ler arquivos, organizar dados e executar operações em DataFrames.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg, sum, count

spark = SparkSession.builder \
    .appName("AgrupamentosAgregacoesDataFrame") \
    .getOrCreate()
```

## Leitura do arquivo de dados

Agora vamos carregar um conjunto de dados com informações de produtos e seus valores.

Arquivo utilizado:

`DADOSANO2C6B3S19A2_auxiliar.csv`

O objetivo é usar esses dados para praticar agrupamentos por categoria e calcular métricas como quantidade, soma e média.

```python
df = spark.read.csv(
    "DADOSANO2C6B3S19A2_auxiliar.csv",
    header=True,
    inferSchema=True
)

df.show()
```

## Agrupamento com `groupBy()` e contagem com `count()`

O método `groupBy()` organiza os dados em grupos a partir de uma coluna.

Neste exemplo, vamos agrupar os produtos pela coluna `categoria` e contar quantos registros existem em cada grupo.

```python
df.groupBy("categoria").count().show()
```

## Soma dos valores com `sum()`

Depois de agrupar os dados, podemos somar os valores de cada grupo.

Neste exemplo, vamos calcular o valor total dos produtos em cada categoria.

```python
df.groupBy("categoria").sum("valor").show()
```

## Média dos valores com `avg()`

Também podemos calcular a média dos valores em cada grupo.

Neste exemplo, vamos descobrir a média de valor dos produtos por categoria.

```python
df.groupBy("categoria").avg("valor").show()
```

## Múltiplas agregações com `agg()`

Quando precisamos analisar mais de uma métrica ao mesmo tempo, utilizamos `agg()`.

Neste exemplo, vamos calcular a média, a soma e a quantidade de valores para cada categoria em uma única operação.

```python
df.groupBy("categoria").agg(
    avg("valor"),
    sum("valor"),
    count("valor")
).show()
```
