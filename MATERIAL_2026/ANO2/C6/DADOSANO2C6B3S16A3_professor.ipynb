## Objetivo prático

Aplicar a criação e a inspeção de DataFrames em PySpark local, utilizando:

- `spark.range()` para criar o DataFrame
- `withColumn()` para criar colunas
- `show()` para visualizar os dados
- `printSchema()` para inspecionar a estrutura (schema)

### Etapa 1: SparkSession

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = (
    SparkSession.builder
    .appName("S16_A3_Prática_DataFrames")
    .master("local[*]")
    .getOrCreate()
)

spark.sparkContext.setLogLevel("ERROR")
print("SparkSession iniciada com sucesso.")
```

### Etapa 2: DataFrame com range — Resposta

```python
df = spark.range(101, 111)  # id de 101 a 110
```

### Verificação rápida — Resposta

```python
df.show(5)
df.count()
```

### Etapa 3: Criando colunas — Resposta

```python
df = (
    df
    .withColumn("nome", F.concat(F.lit("Participante "), F.col("id")))
    .withColumn("turno", F.lit("Manhã"))
    .withColumn("nota", ((F.col("id") % 10) * 0.7 + 3).cast("double"))
)
```

### Etapa 4: show() — Resposta

```python
df.show()
```

### Etapa 5: printSchema() — Resposta

```python
df.printSchema()
```

## Interpretação — Expectativa de resposta

1) `show()` apresenta as **linhas e valores** do DataFrame em formato de tabela.  
2) `printSchema()` apresenta a **estrutura** do DataFrame (schema), com nomes e **tipos** das colunas.  
3) A coluna `nota` fica `double` porque foi **forçada** com `.cast("double")` para permitir cálculos com decimal.  
4) Se `nota` fosse `string`, operações matemáticas poderiam falhar ou gerar resultados incorretos (comparações e somas não funcionam como esperado).
