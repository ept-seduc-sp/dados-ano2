## Objetivo

Compreender como calcular diferenças entre datas e criar colunas temporais simples em DataFrames no PySpark.

Durante a atividade vamos utilizar as seguintes funções:

- `datediff()`
- `current_date()`
- `year()`
- `month()`


## Preparando o ambiente

Antes de trabalhar com DataFrames no PySpark, precisamos iniciar uma sessão Spark.

A `SparkSession` é o ponto de entrada para trabalhar com dados no PySpark.


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date, datediff, current_date, year, month

spark = SparkSession.builder \
    .appName("AnaliseTemporal") \
    .getOrCreate()
```

## Leitura do arquivo de dados

Agora vamos carregar um conjunto de dados com informações de cadastro de clientes.

Arquivo utilizado:

`DADOSANO2C6B3S20A2_auxiliar.csv`

O objetivo é trabalhar com a coluna de data para criar indicadores temporais e calcular a diferença em dias em relação à data atual.


```python
df = spark.read.csv(
    "DADOSANO2C6B3S20A2_auxiliar.csv",
    header=True,
    inferSchema=True
)

df.show()
```

## Conversão da coluna de data

Antes de calcular diferenças entre datas, precisamos garantir que a coluna esteja no tipo correto.

A função `to_date()` converte a coluna de texto para o tipo data.


```python
df = df.withColumn("data_cadastro_convertida", to_date(col("data_cadastro")))

df.show()
```

## Data atual com `current_date()`

A função `current_date()` retorna a data atual do sistema.

Ela pode ser usada para comparar uma data já existente com o momento atual.


```python
df = df.withColumn("data_atual", current_date())

df.show()
```

## Diferença entre datas com `datediff()`

A função `datediff()` calcula a diferença entre duas datas e retorna o resultado em dias.

Neste exemplo, vamos calcular quantos dias se passaram desde o cadastro de cada cliente.


```python
df = df.withColumn(
    "dias_desde_cadastro",
    datediff(col("data_atual"), col("data_cadastro_convertida"))
)

df.show()
```

## Criação de colunas temporais

Também podemos extrair partes da data para organizar melhor a análise.

As funções `year()` e `month()` ajudam a criar colunas com o ano e o mês do cadastro.


```python
df = df.withColumn("ano_cadastro", year(col("data_cadastro_convertida")))
df = df.withColumn("mes_cadastro", month(col("data_cadastro_convertida")))

df.show()
```
