## Objetivo
Compreender como aplicar filtros em um DataFrame com PySpark para selecionar apenas os registros que atendem a critérios específicos.

Durante a atividade vamos utilizar as seguintes funções e recursos:

• `filter()`  
• `where()`  
• condições com `&` e `|`  
• `isin()`  

Essas operações são muito utilizadas na preparação e análise de dados, pois ajudam a focar apenas nas informações relevantes para cada necessidade.


## Preparando o ambiente

Antes de trabalhar com DataFrames no PySpark, precisamos iniciar uma sessão Spark.

A `SparkSession` é o ponto de entrada para trabalhar com dados no PySpark.


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder \
    .appName("FiltrosCondicoesDataFrame") \
    .getOrCreate()
```

## Leitura do arquivo de dados

Agora vamos carregar um conjunto de dados com informações de estudantes.

Arquivo utilizado:

`DADOSANO2C6B3S19A1_auxiliar.csv`

O objetivo é usar esses dados para praticar filtros simples, filtros com múltiplas condições e seleção com `isin()`.


```python
df = spark.read.csv(
    "DADOSANO2C6B3S19A1_auxiliar.csv",
    header=True,
    inferSchema=True
)

df.show()
```

## Filtro simples com `filter()`

O método `filter()` seleciona apenas as linhas que atendem a uma condição.

Neste exemplo, vamos filtrar os estudantes com idade maior ou igual a 18 anos.


```python
df.filter(col("idade") >= 18).show()
```

## Filtro simples com `where()`

O método `where()` tem a mesma função que `filter()`.

Aqui vamos selecionar apenas os estudantes do turno da manhã.


```python
df.where(col("turno") == "manhã").show()
```

## Filtro com múltiplas condições

Podemos combinar condições para deixar a seleção mais específica.

Neste exemplo, vamos buscar estudantes com idade maior ou igual a 18 anos **e** que moram em Campinas.


```python
df.filter(
    (col("idade") >= 18) & (col("cidade") == "Campinas")
).show()
```

## Filtro com mais de uma possibilidade

Também podemos usar o operador `|`, que significa **ou**.

Neste exemplo, vamos selecionar estudantes que estudam no turno da manhã **ou** da tarde.


```python
df.filter(
    (col("turno") == "manhã") | (col("turno") == "tarde")
).show()
```

## Seleção de vários valores com `isin()`

Quando queremos filtrar vários valores possíveis de uma coluna, usamos `isin()`.

Neste exemplo, vamos selecionar apenas os estudantes das cidades de Campinas, São Paulo e Santos.


```python
df.filter(
    col("cidade").isin("Campinas", "São Paulo", "Santos")
).show()
```
