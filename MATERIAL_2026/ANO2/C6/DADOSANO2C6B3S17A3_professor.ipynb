### Etapa 1: Ambiente pronto (SparkSession)


```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("S17A3_Atividade_Professor")
    .master("local[*]")
    .getOrCreate()
)

csv_path = "DADOSANO2C6B3S17A3_auxiliar.csv"

print("SparkSession ativa")
```

### Etapa 3.1: `read.csv()` — Resposta


```python
df_csv_simples = spark.read.csv(csv_path)

df_csv_simples.show(5, truncate=False)
```

### Etapa 3.2: `options` e `inferSchema` — Resposta


```python
df_csv = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .option("sep", ",")
    .csv(csv_path)
)

df_csv.show(5, truncate=False)
```

### Etapa 4: Comparando os schemas — Resposta


```python
print("Schema da leitura simples:")
df_csv_simples.printSchema()

print("\nSchema da leitura com opções:")
df_csv.printSchema()
```

### Etapa 3.3: `show()` e `describe()` — Resposta


```python
print("Amostra do DataFrame com opções:")
df_csv.show(5, truncate=False)

print("\nDescribe do DataFrame com opções:")
df_csv.describe().show()
```
