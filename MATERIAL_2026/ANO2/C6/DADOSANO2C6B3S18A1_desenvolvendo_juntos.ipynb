## Objetivo
Compreender como modificar e criar colunas em um DataFrame durante o processo de preparação de dados.

Durante a atividade vamos utilizar as seguintes funções:

• withColumn()  
• withColumnRenamed()  
• col()  
• lit()

Essas operações são muito utilizadas em pipelines de processamento de dados, pois permitem preparar os dados antes de análises ou modelos.


## Preparando o ambiente

Antes de trabalhar com DataFrames no PySpark, precisamos iniciar uma sessão Spark.

A SparkSession é o ponto de entrada para trabalhar com dados no PySpark.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit

spark = SparkSession.builder \
    .appName("TransformacoesDataFrame") \
    .getOrCreate()
```

## Leitura do arquivo de dados

Agora vamos carregar um pequeno conjunto de dados que representa vendas de produtos.

Arquivo utilizado:

DADOSANO2C6B3S18A1_auxiliar.csv

O objetivo é utilizar esses dados para aplicar transformações em colunas.

```python
df = spark.read.csv(
    "DADOSANO2C6B3S18A1_auxiliar.csv",
    header=True,
    inferSchema=True
)

df.show()
```

## Criando uma nova coluna com withColumn()

O método withColumn() permite criar uma nova coluna ou modificar uma coluna existente.

Neste exemplo vamos criar a coluna **valor_total**, multiplicando:

quantidade × preco_unitario

```python
df = df.withColumn(
    "valor_total",
    col("quantidade") * col("preco_unitario")
)

df.show()
```

## Renomeando colunas

O método withColumnRenamed() permite alterar o nome de uma coluna existente.

Isso é útil quando queremos padronizar nomes ou melhorar a compreensão do conjunto de dados.


```python
df = df.withColumnRenamed(
    "produto",
    "nome_produto"
)

df.show()
```

## Criando colunas com valores constantes

A função lit() permite adicionar uma coluna com um valor fixo para todas as linhas do DataFrame.

Isso pode ser usado, por exemplo, para registrar o ano da coleta dos dados ou a região de origem.


```python
df = df.withColumn(
    "ano_referencia",
    lit(2024)
)

df.show()
```
