### Etapa 1: Ambiente pronto

A SparkSession já está criada para você focar na atividade.


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date, year, month, datediff, current_date, date_format

spark = (
    SparkSession.builder
    .appName("S20A3_Atividade_Professor")
    .master("local[*]")
    .getOrCreate()
)

csv_path = "DADOSANO2C6B3S20A3_auxiliar.csv"

print("SparkSession ativa")
```

### Etapa 2: Leitura do arquivo

Leitura com `header=True` e `inferSchema=True`.

Comentário:
O estudante deve perceber que essas opções ajudam o Spark a interpretar corretamente os nomes das colunas e os tipos de dados.


```python
df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(csv_path)
)

df.show(8)
df.printSchema()
```

### Etapa 3: As colunas data_pedido e data_entrega precisam estar no tipo data para permitir cálculos corretos.

Comentário:
- observar o tipo das colunas de data no schema;
- aplicar funções temporais para organizar e analisar os dados;



```python
df = df.withColumn("pedido_convertido", to_date(col("data_pedido")))
df = df.withColumn("entrega_convertida", to_date(col("data_entrega")))

df.show()
```

### Etapa 4: Colunas temporais com `year()` e `month()`

Comentário:
O objetivo é mostrar que uma data pode gerar outras informações úteis para análise, como ano e mês.


```python
df = df.withColumn("ano_pedido", year(col("pedido_convertido")))
df = df.withColumn("mes_pedido", month(col("pedido_convertido")))

df.show()
```

### Etapa 5: Diferença entre datas com `datediff()`

Comentário:
O estudante deve perceber que `datediff()` retorna a diferença em dias entre duas datas e permite medir intervalos entre eventos.


```python
df = df.withColumn(
    "dias_entrega",
    datediff(col("entrega_convertida"), col("pedido_convertido"))
)

df.show()
```

### Etapa 6: Data atual com `current_date()`

Comentário:
O uso de `current_date()` permite criar análises dinâmicas, comparando uma data registrada com o dia atual.


```python
df = df.withColumn("data_atual", current_date())
df = df.withColumn(
    "dias_desde_pedido",
    datediff(col("data_atual"), col("pedido_convertido"))
)

df.show()
```

### Etapa 7: Formatação com `date_format()`

Comentário:
O objetivo é destacar que a formatação altera a exibição da data, o que é útil para relatórios e apresentações.


```python
df = df.withColumn(
    "pedido_formatado",
    date_format(col("pedido_convertido"), "dd/MM/yyyy")
)

df.show()
```

### Etapa 8: Seleção final para análise

Comentário:
Nesta etapa, o estudante organiza o resultado final, selecionando apenas as colunas mais relevantes para leitura e interpretação.


```python
df.select(
    "cliente",
    "cidade",
    "pedido_formatado",
    "ano_pedido",
    "mes_pedido",
    "dias_entrega",
    "dias_desde_pedido"
).show()
```
