#### **Exercício 1: Avaliação de Satisfação de Clientes**

A empresa TechService lançou um novo suporte técnico 24h e quer saber o nível de satisfação dos clientes. Após um mês, 45 clientes responderam uma pesquisa dando notas de 0 a 10. A média foi 8,2, com desvio padrão de 1,1. A gerência quer um intervalo de confiança de 95% para a satisfação média, ajudando a decidir melhorias futuras.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

media = 8.2
desvio = 1.1
n = 45
confianca = 0.95
t = stats.t.ppf((1 + confianca) / 2, df=n-1)
erro = desvio / np.sqrt(n)
lim_inf, lim_sup = media - t * erro, media + t * erro

plt.figure()
x = np.linspace(lim_inf, lim_sup, 100)
y = stats.t.pdf((x - media) / erro, df=n-1)
plt.plot(x, y)
plt.axvline(lim_inf, color='red', label='Limite Inferior')
plt.axvline(lim_sup, color='green', label='Limite Superior')
plt.axvline(media, color='black', label='Média')
plt.legend()
plt.title("Intervalo de Confiança para Satisfação de Clientes")
plt.show()
print(f"Intervalo de Confiança: ({lim_inf:.2f}, {lim_sup:.2f})")
```

#### **Exercício 2: Avaliação de Pesos em uma Academia**

A academia FitLife quer avaliar o peso médio dos seus alunos após lançar um programa de reeducação alimentar. Dos 50 alunos que participaram, o peso médio foi 72 kg, com desvio padrão de 5 kg. A administração quer saber, com 99% de confiança, o peso médio dos alunos após o programa para avaliar seu sucesso.

```python
media = 72
sigma = 5
n = 50
confianca = 0.99
z = stats.norm.ppf((1 + confianca) / 2)
erro = sigma / np.sqrt(n)
lim_inf, lim_sup = media - z * erro, media + z * erro

plt.figure()
x = np.linspace(lim_inf, lim_sup, 100)
y = stats.norm.pdf(x, media, erro)
plt.plot(x, y)
plt.axvline(lim_inf, color='red', label='Limite Inferior')
plt.axvline(lim_sup, color='green', label='Limite Superior')
plt.axvline(media, color='black', label='Média')
plt.legend()
plt.title("Intervalo de Confiança para Peso Médio")
plt.show()
print(f"Intervalo de Confiança: ({lim_inf:.2f}, {lim_sup:.2f})")
```

#### **Exercício 3: Estimativa de Tempo de Entrega de Pedidos**

A FastTrack Entregas quer melhorar seu serviço e precisa estimar o tempo médio de entrega. Em 35 pedidos analisados, o tempo médio foi de 50 minutos, com desvio padrão de 8 minutos. Eles precisam de um intervalo de confiança de 90% para garantir aos clientes um prazo confiável.

```python
media = 50
desvio = 8
n = 35
confianca = 0.90
t = stats.t.ppf((1 + confianca) / 2, df=n-1)
erro = desvio / np.sqrt(n)
lim_inf, lim_sup = media - t * erro, media + t * erro

plt.figure()
x = np.linspace(lim_inf, lim_sup, 100)
y = stats.t.pdf((x - media) / erro, df=n-1)
plt.plot(x, y)
plt.axvline(lim_inf, color='red', label='Limite Inferior')
plt.axvline(lim_sup, color='green', label='Limite Superior')
plt.axvline(media, color='black', label='Média')
plt.legend()
plt.title("Intervalo de Confiança para Tempo de Entrega")
plt.show()
print(f"Intervalo de Confiança: ({lim_inf:.2f}, {lim_sup:.2f})")
```
