#### **Exercício 1: Estimativa de Desempenho de Alunos**

Uma escola quer avaliar o desempenho médio de alunos no ENEM para aprimorar seu plano pedagógico. Após analisar o histórico, selecionou-se uma amostra de 60 alunos, cuja pontuação média foi 680 pontos, com desvio padrão de 45 pontos. O coordenador deseja um intervalo de confiança de 99% para estimar a média real de todos os alunos da escola, a fim de comparar o desempenho com outras instituições e planejar intervenções.



```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

media = 680
desvio = 45
n = 60
confianca = 0.99
t = stats.t.ppf((1 + confianca) / 2, df=n-1)
erro = desvio / np.sqrt(n)
lim_inf, lim_sup = media - t * erro, media + t * erro

plt.figure(figsize=(8,5))
x = np.linspace(lim_inf, lim_sup, 100)
y = stats.t.pdf((x - media) / erro, df=n-1)
plt.plot(x, y, color='blue')
plt.axvline(lim_inf, color='red', linestyle='--', label='Limite Inferior')
plt.axvline(lim_sup, color='green', linestyle='--', label='Limite Superior')
plt.axvline(media, color='black', label='Média Amostral')
plt.legend()
plt.title("Intervalo de Confiança para Desempenho no ENEM")
plt.show()

print(f"Intervalo de Confiança: ({lim_inf:.2f}, {lim_sup:.2f})")
```

#### **Exemplo 2: Análise de Custos em Construção Civil**

Uma construtora quer estimar o custo médio por metro quadrado de projetos para definir orçamentos futuros. A análise de 55 projetos revelou um custo médio de R$ 2.800/m², com desvio padrão populacional de R$ 300/m². A empresa precisa de um intervalo de confiança de 95% para o custo médio real, a fim de negociar com fornecedores e prever margens de lucro com maior precisão.

```python
media = 2800
sigma = 300
n = 55
confianca = 0.95
z = stats.norm.ppf((1 + confianca) / 2)
erro = sigma / np.sqrt(n)
lim_inf, lim_sup = media - z * erro, media + z * erro

plt.figure(figsize=(8,5))
x = np.linspace(lim_inf, lim_sup, 100)
y = stats.norm.pdf(x, media, erro)
plt.plot(x, y, color='blue')
plt.axvline(lim_inf, color='red', linestyle='--', label='Limite Inferior')
plt.axvline(lim_sup, color='green', linestyle='--', label='Limite Superior')
plt.axvline(media, color='black', label='Média Amostral')
plt.legend()
plt.title("Intervalo de Confiança para Custos de Construção")
plt.show()

print(f"Intervalo de Confiança: ({lim_inf:.2f}, {lim_sup:.2f})")
```

#### **Exemplo 3: Avaliação de Eficiência Energética**

Após instalar painéis solares, uma empresa de energia deseja estimar o consumo médio de eletricidade em residências. Foram analisadas 40 casas, com consumo médio de 350 kWh/mês e desvio padrão de 25 kWh. O objetivo é calcular um intervalo de confiança de 90% para garantir aos novos clientes que a instalação resultará em economia real de energia.

```python
media = 350
desvio = 25
n = 40
confianca = 0.90
t = stats.t.ppf((1 + confianca) / 2, df=n-1)
erro = desvio / np.sqrt(n)
lim_inf, lim_sup = media - t * erro, media + t * erro

plt.figure(figsize=(8,5))
x = np.linspace(lim_inf, lim_sup, 100)
y = stats.t.pdf((x - media) / erro, df=n-1)
plt.plot(x, y, color='blue')
plt.axvline(lim_inf, color='red', linestyle='--', label='Limite Inferior')
plt.axvline(lim_sup, color='green', linestyle='--', label='Limite Superior')
plt.axvline(media, color='black', label='Média Amostral')
plt.legend()
plt.title("Intervalo de Confiança para Consumo de Energia")
plt.show()

print(f"Intervalo de Confiança: ({lim_inf:.2f}, {lim_sup:.2f})")
```
