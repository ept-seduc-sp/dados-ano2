#### Tarefa 1: Simulação de uma Distribuição Normal

**Enunciado:** 
Simulem um conjunto de dados representando a altura de uma população de 1000 indivíduos, com média de 1,70m e desvio padrão de 0,10m. Utilize a biblioteca numpy para gerar os dados e matplotlib para criar um histograma e sobrepor a curva normal ajustada.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Gerar dados simulados
np.random.seed(42)
alturas = np.random.normal(loc=1.70, scale=0.10, size=1000)

# Criar histograma
plt.figure(figsize=(8,5))
plt.hist(alturas, bins=30, edgecolor='black', alpha=0.7, density=True)

# Criar curva normal ajustada
x = np.linspace(min(alturas), max(alturas), 100)
y = norm.pdf(x, loc=1.70, scale=0.10)
plt.plot(x, y, color='red', linewidth=2, label='Curva Normal Ajustada')

plt.title('Distribuição Normal das Alturas')
plt.xlabel('Altura (m)')
plt.ylabel('Densidade')
plt.legend()
plt.grid()
plt.show()
```

#### Perguntas:

   
1. **O histograma tem formato simétrico? O que isso significa?**  
   - Sim, o histograma tem formato simétrico, indicando que os valores estão distribuídos de maneira equilibrada em torno da média. Isso sugere que os dados seguem uma distribuição normal, onde a média, mediana e moda coincidem.

2. **Como seria a distribuição se o desvio padrão fosse maior?**  
   - Se o desvio padrão fosse maior, os dados estariam mais espalhados ao redor da média, tornando a curva da distribuição mais achatada e larga. Isso significa maior variabilidade nas alturas da população simulada.

3. **Qual a interpretação da média dos dados gerados?**  
   - A média representa o valor central da distribuição. No caso dos dados simulados, uma média de 1,70m indica que a maioria das alturas está próxima desse valor, com algumas variações ao redor.

4. **Como a curva normal ajustada ajuda na interpretação dos dados?**  
   - A curva normal ajustada fornece uma referência visual sobre a distribuição dos dados. Se o histograma estiver bem ajustado à curva, significa que os dados realmente seguem uma distribuição normal. Caso contrário, pode haver outliers ou assimetrias na amostra.



#### Tarefa 2: Probabilidades na Distribuição Normal

**Enunciado:**

Calculem as seguintes probabilidades utilizando a função de distribuição acumulada da distribuição normal (scipy.stats.norm.cdf):

1. A probabilidade de um indivíduo ter altura entre 1,60m e 1,80m.

2. A probabilidade de um indivíduo ter altura menor que 1,60m (probabilidade à esquerda).

3. A probabilidade de um indivíduo ter altura maior que 1,80m (probabilidade à direita).

4. A probabilidade de um indivíduo ter altura fora do intervalo de 1,60m a 1,80m (probabilidade bilateral).

```python
from scipy.stats import norm

# Definir parâmetros da distribuição
media = 1.70
desvio_padrao = 0.10

# Probabilidade entre 1,60m e 1,80m
probabilidade_entre = norm.cdf(1.80, loc=media, scale=desvio_padrao) - norm.cdf(1.60, loc=media, scale=desvio_padrao)

# Probabilidade menor que 1,60m (esquerda)
probabilidade_esquerda = norm.cdf(1.60, loc=media, scale=desvio_padrao)

# Probabilidade maior que 1,80m (direita)
probabilidade_direita = 1 - norm.cdf(1.80, loc=media, scale=desvio_padrao)

# Probabilidade bilateral (fora do intervalo 1,60m a 1,80m)
probabilidade_bilateral = probabilidade_esquerda + probabilidade_direita

print(f'Probabilidade entre 1,60m e 1,80m: {probabilidade_entre:.4f}')
print(f'Probabilidade menor que 1,60m: {probabilidade_esquerda:.4f}')
print(f'Probabilidade maior que 1,80m: {probabilidade_direita:.4f}')
print(f'Probabilidade fora do intervalo (1,60m a 1,80m): {probabilidade_bilateral:.4f}')
```

#### Perguntas

1. **Como a probabilidade encontrada para o intervalo 1,60m - 1,80m pode ser interpretada no contexto real?**  
   - A probabilidade encontrada representa a fração da população que tem altura entre 1,60m e 1,80m. Se a probabilidade for, por exemplo, **95,4%**, significa que **95,4% das pessoas dessa população têm altura nesse intervalo**. Isso mostra a proporção de indivíduos dentro da faixa considerada "normal" em relação à distribuição estudada.

2. **O que aconteceria com as probabilidades se o desvio padrão fosse reduzido para 0,05m?**  
   - Com um desvio padrão menor, a distribuição dos dados ficaria **mais concentrada em torno da média** (1,70m). Isso significa que **a probabilidade de um indivíduo estar entre 1,60m e 1,80m aumentaria**, pois haveria menos variação nos dados.  
   - Como consequência, as probabilidades laterais (menor que 1,60m e maior que 1,80m) diminuiriam, pois haveria menos valores distantes da média.

3. **Em quais situações reais pode ser interessante calcular probabilidades laterais (menor que um valor ou maior que outro valor)?**  
   - Em processos industriais para verificar a quantidade de produtos defeituosos que estão abaixo ou acima do padrão esperado.  
   - Na área da saúde, para analisar a proporção de pacientes com **pressão arterial muito baixa ou muito alta**.  
   - Em exames educacionais, para identificar a fração de alunos com **notas muito baixas ou muito altas** em relação à média da turma.  
   - No mercado financeiro, para avaliar o risco de um investimento ter retornos extremos para **valores abaixo ou acima de um limite crítico**.

4. **Quando faz sentido utilizar a probabilidade bilateral?**  
   - Quando se deseja medir a proporção de elementos que **se desviam significativamente da média**, ou seja, valores **muito abaixo ou muito acima** do esperado.  
   - Em controle de qualidade, para calcular a taxa de produtos fora das especificações.  
   - Em estatísticas médicas, para analisar pacientes com **valores de glicose muito altos ou muito baixos** em relação ao padrão saudável.  
   - Em previsões climáticas, para determinar a probabilidade de temperaturas extremas **(muito frias ou muito quentes)** em relação à média histórica.


#### Reflexões e discussão com a sala

1. **Em quais cenários reais a distribuição normal pode não ser um bom modelo?**  
   - Quando os dados possuem **assimetria** ou caudas longas, como distribuições **exponenciais** (tempo de vida de equipamentos) ou **log-normais** (renda de uma população).  
   - Quando há **valores extremos** (outliers), que podem distorcer a normalidade dos dados.  
   - Em distribuições **discretas**, como número de clientes em uma loja por dia, onde a normalidade não se aplica diretamente.  
   - Quando há um **limite inferior ou superior**, como a distribuição de idades de uma população (não pode haver valores negativos).  

2. **Como a alteração da média e do desvio padrão impacta a forma da distribuição?**  
   - **Alteração da média (μ):** Desloca a distribuição para a direita (se aumentar) ou para a esquerda (se diminuir), sem modificar sua forma.  
   - **Alteração do desvio padrão (σ):** Um desvio padrão maior **alarga** a distribuição, tornando-a mais dispersa. Um desvio padrão menor **estreita** a distribuição, concentrando mais valores próximos da média.  

3. **Que outras variáveis do dia a dia poderiam ser modeladas com a distribuição normal?**  
   - **Notas de alunos** em um exame padronizado.  
   - **Altura e peso** de uma população adulta.  
   - **Temperaturas diárias** em uma cidade ao longo do ano.  
   - **Variações na produção industrial**, como o peso de um produto manufaturado.  
   - **Flutuações em retornos financeiros**, considerando mercados relativamente estáveis.

#### Tarefa 1: Cálculo de Probabilidades na Distribuição Normal

**Enunciado:**

1. Um conjunto de dados representa alturas de uma população com média de 1,70m e desvio padrão de 0,10m. Qual a probabilidade de um indivíduo ter altura menor que 1,65m?

2. Qual a probabilidade de um indivíduo ter altura entre 1,65m e 1,80m?

3. Qual a probabilidade de um indivíduo ter altura maior que 1,80m?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir média e desvio padrão
media = 1.70
desvio_padrao = 0.10

# Probabilidade de altura menor que 1,65m
prob_menor_165 = norm.cdf(1.65, loc=media, scale=desvio_padrao)

# Probabilidade de altura entre 1,65m e 1,80m
prob_entre_165_180 = norm.cdf(1.80, loc=media, scale=desvio_padrao) - norm.cdf(1.65, loc=media, scale=desvio_padrao)

# Probabilidade de altura maior que 1,80m
prob_maior_180 = 1 - norm.cdf(1.80, loc=media, scale=desvio_padrao)

print(f'P(Altura < 1,65m): {prob_menor_165:.4f}')
print(f'P(1,65m < Altura < 1,80m): {prob_entre_165_180:.4f}')
print(f'P(Altura > 1,80m): {prob_maior_180:.4f}')
```

### **Perguntas**

1. **Como interpretar os valores de probabilidade obtidos?**  
   - Os valores de probabilidade representam a fração da população que atende a determinada condição. Por exemplo, se P(Altura < 1,65m) = 0.3085, significa que **30,85% das pessoas têm altura inferior a 1,65m**.

2. **O que acontece se o desvio padrão for reduzido para 0,05m?**  
   - Se o desvio padrão for menor, os dados estarão **mais concentrados em torno da média**, reduzindo a dispersão. Isso significa que a probabilidade de valores muito distantes da média diminui, tornando a curva mais "alta e estreita".

3. **Em que cenários práticos esse tipo de cálculo pode ser útil?**  
   - Na análise de **qualidade industrial**, para medir quantos produtos estão fora do padrão.  
   - No setor de **saúde**, para identificar quantas pessoas possuem um peso abaixo ou acima do recomendado.  
   - Na **educação**, para calcular quantos alunos tiveram notas dentro de um certo intervalo.

#### Tarefa 2: Simulação e Visualização da Tabela de Probabilidades

**Enunciado:**

Utilizem Python para gerar uma curva da distribuição normal e destacar a área correspondente a uma probabilidade específica. Exemplo: Probabilidade de altura menor que 1,65m.

```python
import numpy as np
import matplotlib.pyplot as plt

# Criar valores para a curva normal
x = np.linspace(1.40, 2.00, 1000)
y = norm.pdf(x, loc=media, scale=desvio_padrao)

# Criar gráfico
plt.figure(figsize=(8,5))
plt.plot(x, y, label='Distribuição Normal', color='blue')
plt.fill_between(x, y, where=(x<=1.65), color='red', alpha=0.5, label='P(Altura < 1,65m)')
plt.axvline(1.65, color='red', linestyle='--', label='x = 1,65m')

plt.title('Distribuição Normal das Alturas')
plt.xlabel('Altura (m)')
plt.ylabel('Densidade')
plt.legend()
plt.grid()
plt.show()
```

#### Simulando a tabela de distribuição normal

```python
from scipy.stats import norm

# Definir média e desvio padrão
media = 1.70
desvio_padrao = 0.10

# Cálculo de probabilidades usando a tabela de distribuição normal (Z-table)
# Convertendo valores para Z-score
z_165 = (1.65 - media) / desvio_padrao
z_180 = (1.80 - media) / desvio_padrao

# Probabilidade de altura menor que 1,65m
prob_menor_165 = norm.cdf(z_165)

# Probabilidade de altura entre 1,65m e 1,80m
prob_entre_165_180 = norm.cdf(z_180) - norm.cdf(z_165)

# Probabilidade de altura maior que 1,80m
prob_maior_180 = 1 - norm.cdf(z_180)

prob_menor_165, prob_entre_165_180, prob_maior_180
```

Cálculos das **tarefas 1 e 2**, utilizando a **tabela de distribuição normal (Z-table)**:

1. **Probabilidade de um indivíduo ter altura menor que 1,65m**  
   - **Valor de Z:** $ Z = \frac{1.65 - 1.70}{0.10} = -0.5 $  
   - **Probabilidade acumulada (P(X < 1.65)):** **0.3085** (30,85%)

2. **Probabilidade de um indivíduo ter altura entre 1,65m e 1,80m**  
   - **Valores de Z:** $ Z_{1.65} = -0.5 $, $ Z_{1.80} = 1.0 $  
   - **P(1.65 < X < 1.80):** **0.5328** (53,28%)

3. **Probabilidade de um indivíduo ter altura maior que 1,80m**  
   - **Valor de Z:** $ Z = \frac{1.80 - 1.70}{0.10} = 1.0 $  
   - **P(X > 1.80):** **0.1587** (15,87%)

#### Exercício 1 – Controle de Qualidade em uma Indústria

**Enunciado:**

Uma fábrica produz parafusos com um diâmetro médio de 5mm e um desvio padrão de 0,02mm. O controle de qualidade rejeita parafusos com diâmetro menor que 4,96mm ou maior que 5,04mm. Qual a porcentagem de parafusos que serão rejeitados?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir parâmetros
media = 5.00
desvio_padrao = 0.02
limite_inferior = 4.96
limite_superior = 5.04

# Calcular probabilidades
prob_inferior = norm.cdf(limite_inferior, loc=media, scale=desvio_padrao)
prob_superior = 1 - norm.cdf(limite_superior, loc=media, scale=desvio_padrao)
prob_rejeitados = prob_inferior + prob_superior

print(f'Porcentagem de parafusos rejeitados: {prob_rejeitados * 100:.2f}%')
```

#### Exercício 2 – Notas de um Exame Escolar

**Enunciado:**

As notas de um exame seguem uma distribuição normal com média 70 pontos e desvio padrão de 10 pontos. Um aluno precisa de pelo menos 85 pontos para ser aprovado com mérito. Qual a probabilidade de um aluno ser aprovado com mérito?

```python
# Definir parâmetros
media = 70
desvio_padrao = 10
nota_corte = 85

# Calcular probabilidade
prob_aprovacao = 1 - norm.cdf(nota_corte, loc=media, scale=desvio_padrao)

print(f'Probabilidade de aprovação com mérito: {prob_aprovacao * 100:.2f}%')
```
